```python
from llama_index.core import (
    VectorStoreIndex,
    SimpleDirectoryReader
)
```

```python
# LOAD PDF DOCUMENTS
documents = SimpleDirectoryReader("data").load_data()

print(f"\nLoaded {len(documents)} document chunks\n")
```

```python
# (OPTIONAL): Lets see the chunks
from llama_index.core.node_parser import SentenceSplitter

splitter = SentenceSplitter(
    chunk_size=200,      # characters/tokens approx
    chunk_overlap=20
)

# -----------------------------------
# CREATE CHUNKS
# -----------------------------------

nodes = splitter.get_nodes_from_documents(documents)

# -----------------------------------
# SHOW TOTAL CHUNKS
# -----------------------------------

print(f"\nTotal Chunks Created: {len(nodes)}\n")

# -----------------------------------
# DISPLAY CHUNKS
# -----------------------------------

for i, node in enumerate(nodes):

    chunk_text = node.text

    print("=" * 80)
    print(f"CHUNK NUMBER: {i+1}")
    print("=" * 80)

    print("\nChunk Text:\n")
    print(chunk_text)

    print("\nChunk Length:")
    print(len(chunk_text))

    print("\nMetadata:")
    print(node.metadata)

    print("\n")
```

```python
# ideally should be in .env file
import os

os.environ["OPENAI_API_KEY"] = "key_here"
```

```python
# CREATE VECTOR INDEX
index = VectorStoreIndex.from_documents(documents)

print("Vector Index Created Successfully\n")
```

```python
# CREATE QUERY ENGINE
query_engine = index.as_query_engine()
print("Query Engine Ready\n")

# ASK QUESTIONS
while True:

    question = input("\nAsk a question (type 'q' to quit): ")

    if question.lower() == "q":
        break

    response = query_engine.query(question)

    print("\nAnswer:")
    print(response)
```
