## 2. Basic API Demo (Text Generation)

```python
from openai import OpenAI

# Initialize client
# NOT A GOOD PRACTISE TO PUT KEY HERE
API_KEY="key_here" 
client = OpenAI(api_key=API_KEY)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "Explain me role of API in LLM in less than 50 words."}
    ],
    temperature=0.7
)
```

```python
# 1) View the complete response

print(response)
```

```python
# 2) Print the relevant response

print(response.choices[0].message.content)
```

# 4. The chat bot

```python
system_prompt = """
You are a polite customer support assistant for an e-commerce company.

Rules:
- Be polite and professional
- Ask for order ID when needed
- Handle refunds, complaints, and product queries
"""

messages = [
    {"role": "system", "content": system_prompt}
]

while True:
    # Example: I ordered shoes. I need to return shoes.
    user_input = input("You (type exit to quit): ") 
    
    if user_input.lower() == "exit":
        break
    
    messages.append({"role": "user", "content": user_input})
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0.7
    )
    
    reply = response.choices[0].message.content
    print("Bot:", reply)
    
    messages.append({"role": "assistant", "content": reply}) # This way it remember previous response
```

```python
# Now lets view all previous messages
print(messages)
```
