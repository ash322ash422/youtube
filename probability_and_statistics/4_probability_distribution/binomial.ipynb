# Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom # binomial distri 
```

```python
plt.style.use("dark_background")
```

### You toss a coin 5 times and prob of success (getting H) is 0.5
### What is the PMF ?
### What is probability of getting 3 heads ?

```python
# Number of trials (keep constant)
n = 5

# Possible number of successes
x = np.arange(0, n+1) # x = 0, 1, 2, 3, 4, 5

p = 0.5

probabilities = binom.pmf(x, n, p)
probabilities
```

### Probability of getting 0 heads = 0.03125
### Probability of getting 1 heads = 0.15625
### Probability of getting 2 heads = 0.3125
### Probability of getting 3 heads = 0.3125
### Probability of getting 4 heads = 0.15625
### Probability of getting 5 heads = 0.03125

```python
# bar plot: 
plt.style.use("dark_background")
plt.figure(figsize=(5,3))

plt.bar(x, probabilities, color='yellow')

plt.xlabel("Number of Successes")
plt.ylabel("Probability")

plt.tight_layout()
plt.show()
```

```python
# Alternative: stem plot
plt.figure(figsize=(5,3))

plt.stem(x, probabilities)

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.tight_layout()
plt.show()
```

### An assembly line produces machine parts. The probability that a part is defective is 0.2. Three parts are selected randomly and independently.

### What is the probability that exactly 2 parts are defective?


```python
# Number of trials (keep constant)
n = 3

# Possible number of successes
x = np.arange(0, n+1) # x = 0, 1, 2, 3

p = 0.2

probabilities = binom.pmf(x, n, p)
probabilities
```

### Probability of getting 0 defect = 0.512
### Probability of getting 1 defect = 0.384
### Probability of getting 2 defect = 0.096
### Probability of getting 3 defect = 0.008

```python
# plot: 

plt.style.use("dark_background")
plt.figure(figsize=(5,3))

plt.bar(x, probabilities, color='yellow')

plt.xlabel("Number of Successes")
plt.ylabel("Probability")

plt.tight_layout()
plt.show()
```

```python
# Alternative: stem plot

plt.style.use("dark_background")
plt.figure(figsize=(5,3))

plt.stem(x, probabilities)

plt.xlabel("Number of Successes")
plt.ylabel("Probability")

plt.tight_layout()
plt.show()
```
