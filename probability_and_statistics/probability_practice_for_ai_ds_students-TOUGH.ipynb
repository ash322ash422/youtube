### Solution 2.2
Using Bayes' theorem:

$$P(D \mid +) = \frac{P(+ \mid D)\,P(D)}{P(+ \mid D)P(D) + P(+ \mid \lnot D)P(\lnot D)}$$

With $P(D) = 1/500 = 0.002$, $P(+|D) = 0.99$ (sensitivity), and
$P(+|\lnot D) = 1 - \text{specificity} = 0.02$:

Even with a seemingly accurate test, the low base rate of the disease means most positives are
**false positives** — this is a crucial intuition for any imbalanced classification problem
(fraud, rare-event detection, etc.).


```python
P_disease = 1/500
sensitivity = 0.99
specificity = 0.98
P_no_disease = 1 - P_disease
P_pos_given_no_disease = 1 - specificity

numerator = sensitivity * P_disease
denominator = numerator + P_pos_given_no_disease * P_no_disease
P_disease_given_pos = numerator / denominator

print(f"P(disease | positive test) = {P_disease_given_pos:.4f} (~{P_disease_given_pos*100:.1f}%)")
```

### Solution 2.3
Unnormalized scores:
- $\text{score}_{spam} = P(spam) \cdot P(free|spam) \cdot P(meeting|spam)$
- $\text{score}_{not} = P(not) \cdot P(free|not) \cdot P(meeting|not)$

Then normalize: $P(spam \mid \text{evidence}) = \dfrac{\text{score}_{spam}}{\text{score}_{spam}+\text{score}_{not}}$

Note the result: despite "free" being a strong spam signal, "meeting" is a strong *non-spam*
signal, and Naive Bayes multiplies these (naively assumed independent) likelihoods together —
this is exactly how a real Naive Bayes spam filter reasons over many words at once.


```python
P_spam, P_not = 0.4, 0.6
P_free_spam, P_free_not = 0.6, 0.1
P_meet_spam, P_meet_not = 0.05, 0.3

score_spam = P_spam * P_free_spam * P_meet_spam
score_not = P_not * P_free_not * P_meet_not

posterior_spam = score_spam / (score_spam + score_not)
print(f"Score (spam)     = {score_spam:.5f}")
print(f"Score (not spam) = {score_not:.5f}")
print(f"P(spam | evidence) = {posterior_spam:.3f}")
```

### Solution 5.1
(a) $L(p) = p^7 (1-p)^3$ (product of individual Bernoulli probabilities).

(b) Log-likelihood: $\ell(p) = 7\ln p + 3\ln(1-p)$.
Differentiate: $\ell'(p) = \frac{7}{p} - \frac{3}{1-p} = 0 \Rightarrow 7(1-p) = 3p \Rightarrow p = 7/10 = 0.7$.

This matches the intuitive estimator: $\hat p = \text{(number of heads)} / \text{(number of flips)}$
— MLE for a Bernoulli/Binomial parameter is just the sample proportion.


```python
n_heads, n_tails = 7, 3
mle_p = n_heads / (n_heads + n_tails)

ps = np.linspace(0.001, 0.999, 500)
log_likelihood = n_heads*np.log(ps) + n_tails*np.log(1-ps)

plt.plot(ps, log_likelihood)
plt.axvline(mle_p, color='red', linestyle='--', label=f'MLE p_hat = {mle_p}')
plt.xlabel('p')
plt.ylabel('log-likelihood')
plt.legend()
plt.title('Log-likelihood peaks exactly at the analytical MLE')
plt.show()

print("Analytical MLE:", mle_p)
```

(b) The stationary distribution $\pi$ satisfies $\pi T = \pi$. We can find it either by iterating

```python
T = np.array([[0.5, 0.4, 0.1],
              [0.3, 0.4, 0.3],
              [0.2, 0.3, 0.5]])

start = np.array([1, 0, 0])

# (a) distribution after 2 steps
after_2_steps = start @ np.linalg.matrix_power(T, 2)
print("Distribution after 2 steps [H, P, C]:", after_2_steps)

# (b) stationary distribution via power iteration
pi = np.array([1/3, 1/3, 1/3])
for _ in range(1000):
    pi = pi @ T
print("Stationary distribution [H, P, C]:", pi)

# sanity check: pi @ T should equal pi
print("Check pi @ T == pi:", np.allclose(pi @ T, pi))
```

### Solution 5.3
(a) Cross-entropy only involves the term where $y_i = 1$: $-\log(0.2) \approx 1.609$

(b) $H(\hat y) = -(0.7\log 0.7 + 0.2\log 0.2 + 0.1\log 0.1) \approx 0.802$ nats.
A *low* entropy of $\hat y$ means the model is putting most of its probability mass on one class,
i.e. it is confident (though not necessarily correct — confidence and correctness are different!).

(c) With $\hat y = [0.1, 0.8, 0.1]$: cross-entropy $= -\log(0.8) \approx 0.223$, much lower than (a).
This shows how cross-entropy loss sharply penalizes the model for putting low probability on the
*true* class, and rewards concentrating probability mass correctly.


```python
def cross_entropy(y_true, y_pred, eps=1e-12):
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.sum(y_true * np.log(y_pred))

def entropy(p, eps=1e-12):
    p = np.clip(p, eps, 1 - eps)
    return -np.sum(p * np.log(p))

y_true = np.array([0, 1, 0])
y_pred_1 = np.array([0.7, 0.2, 0.1])
y_pred_2 = np.array([0.1, 0.8, 0.1])

print("Cross-entropy with y_pred_1:", cross_entropy(y_true, y_pred_1))
print("Entropy of y_pred_1:        ", entropy(y_pred_1))
print("Cross-entropy with y_pred_2:", cross_entropy(y_true, y_pred_2))
```

### Solution 6.2
The histogram of sample means should look approximately **bell-shaped / Normal**, even though
the original Exponential distribution is heavily right-skewed. This is the **Central Limit
Theorem**: the sampling distribution of the mean tends toward Normal as sample size grows,
*regardless* of the underlying population distribution's shape. This is precisely why we can
apply Normal-theory-based hypothesis tests (like a t-test in A/B testing) to averages of metrics
that are themselves not Normally distributed.


```python
n_samples_per_mean = 30
n_repeats = 10_000

sample_means = [np.random.exponential(scale=1.0, size=n_samples_per_mean).mean()
                for _ in range(n_repeats)]

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].hist(np.random.exponential(scale=1.0, size=10_000), bins=50)
axes[0].set_title("Original Exponential distribution\n(skewed)")

axes[1].hist(sample_means, bins=50)
axes[1].set_title("Distribution of sample means (n=30)\n(approximately Normal - CLT!)")
plt.tight_layout()
plt.show()
```

### Solution 6.3
(a)/(b): both the Monte Carlo estimate and exact formula should give roughly **70.6%**.

(c) You need $n$ such that the collision probability exceeds 50% — from the classic birthday
problem with $m=365$, that threshold is around $n \approx 23$. In hashing terms: if you hash only
23 items into 365 buckets, you already have better than even odds of a collision! This is why
hash tables and feature-hashing schemes need buckets ($m$) that scale roughly like $n^2$, not $n$,
to keep collision probability low — a very practical consequence of this "paradox."


```python
n_people = 30
n_trials = 100_000

def has_shared_birthday(n_people):
    birthdays = np.random.randint(0, 365, size=n_people)
    return len(set(birthdays)) < n_people

# (a) Monte Carlo estimate
results = [has_shared_birthday(n_people) for _ in range(n_trials)]
mc_estimate = np.mean(results)

# (b) exact formula
prob_no_match = np.prod([(365 - i) / 365 for i in range(n_people)])
exact = 1 - prob_no_match

print(f"Monte Carlo estimate (n=30 people): {mc_estimate:.4f}")
print(f"Exact analytical answer:             {exact:.4f}")

# (c) smallest n with >50% collision probability
for n in range(2, 60):
    prob_no_match_n = np.prod([(365 - i) / 365 for i in range(n)])
    if 1 - prob_no_match_n > 0.5:
        print(f"Smallest n with P(collision) > 50%: n = {n}")
        break
```
