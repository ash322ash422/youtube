```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data
conversions = np.array([240, 200])  # B first (important)
users = np.array([1000, 1000])

# Right-tailed test (is B > A?)
z_stat, p_value = proportions_ztest(conversions, users, alternative='larger')

print("Z-stat:", z_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Green button is significantly better")
else:
    print("No significant improvement")
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data (B first, A second)
clicks = np.array([750, 600])   # Model B, Model A
users  = np.array([5000, 5000])

# Right-tailed test: H1 → p_B > p_A
z_stat, p_value = proportions_ztest(clicks, users, alternative='larger')

print("Z-stat:", z_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Model B is significantly better")
else:
    print("No significant improvement")
```
