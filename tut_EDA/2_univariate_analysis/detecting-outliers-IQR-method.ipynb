# Detecting outliers: IQR Method
- Inter Quartile Range 

```python
import pandas as pd

df = pd.read_csv("data_outliers_iqr.csv")
print(df)
```

```python
# Compute IQR

Q1 = df["Score"].quantile(0.25)
Q3 = df["Score"].quantile(0.75)
IQR = Q3 - Q1

print(Q1, Q3)
print(IQR)
```

```python
# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR # min
upper_bound = Q3 + 1.5 * IQR # max

print(lower_bound, upper_bound)
```

```python
# print outliers

mask = (df["Score"] < lower_bound) | (df["Score"] > upper_bound)

outliers = df[mask]
print("Outliers:\n", outliers)
```

# Trimming

```python
# trimming: removing outliers from df

mask = (df["Score"] >= lower_bound) & (df["Score"] <= upper_bound)

df_trimmed = df[mask]

print(df_trimmed)
```

# Capping or winsorization

```python
print(lower_bound, upper_bound)
print("Outliers:\n", outliers)
```

```python
# cap the outliers
df_winsorized = df.copy()

df_winsorized["Score"] = df_winsorized["Score"].clip(lower=lower_bound, upper=upper_bound)

print(df_winsorized)
```
