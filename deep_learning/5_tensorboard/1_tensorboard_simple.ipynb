# Tensorboard

## Basic Workflow (Industry Pattern)

👉 Steps used everywhere:

1) Train model
2) Log metrics (loss, accuracy)
3) Log graphs / histograms / images
4) Launch TensorBoard
5) Compare experiments

```python
import tensorflow as tf
import tensorboard

print(tf.__version__)
print(tensorboard.__version__)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
import datetime
```

```python
# Dummy data
import numpy as np

X = np.random.rand(1000, 20)
y = np.random.randint(0, 10, 1000)

# Log directory
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
# Create simple model
model = tf.keras.Sequential([
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

```

```python
# TensorBoard callback
tensorboard_callback = tf.keras.callbacks.TensorBoard(
    log_dir=log_dir,
    histogram_freq=1
)

# Train model
model.fit(X, y, epochs=5, callbacks=[tensorboard_callback])
```

# Launch TensorBoard

Run following in the same directory that contain this code

> tensorboard --logdir=logs/fit
