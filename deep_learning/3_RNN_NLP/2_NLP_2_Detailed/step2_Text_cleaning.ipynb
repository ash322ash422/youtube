```python
import re
import string


# Example usage
# Realistic multi-line paragraph
original_text = """
Hey @Sarah,      did you check out the new iPhone 14 launch? <br> It’s all over the internet! I think they’ve finally added USB-C.
Yesterday, I         visited https://apple.com to see the     specs, but honestly, it looks the same as iPhone 13.
Also, lol @Mike was like, “It’s basically a glorified iPhone 6s.” 😂
I mean, don't get me wrong—it has a better camera, sure, but $1099? That’s too much.
BTW, have you watched the Netflix show “The Tech Diaries”? It's got 9 episodes in season 2.
There’s a funny scene where a robot says: “Humans are obsolete.” Classic.
Anyway, I’m thinking of switching to Android. The new Pixel seems promising!
Let’s catch up soon—maybe this Saturday at Burgur Palace? 🍔
Oh and I found an article (in <div class='news'>The Verge</div>) claiming battery life is improved.
Talk soon! <p>Cheers, Alex</p>
"""

def to_lowercase(text):
    return text.lower()
print("Original:", original_text)
text = to_lowercase(original_text)
print("Cleaned:\n", text)
print("#######################################")

def remove_user_mentions(text):
    return re.sub(r'@\w+', '', text)
print("Original:", text)
text = remove_user_mentions(text)
print("Cleaned:\n", text)
print("#######################################")


def remove_urls(text):
    return re.sub(r'http\S+|www\.\S+', '', text) # Pattern matches http(s) URLs and www URLs
print("Original:", text)
text = remove_urls(text)
print("Cleaned:\n", text)
print("#######################################")


def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)
print("Original:", text)
text = remove_html_tags(text)
print("Cleaned:\n", text)
print("#######################################")



def remove_extra_whitespace(text):
    return re.sub(r'\s+', ' ', text).strip()
print("Original:", text)
text = remove_extra_whitespace(text)
print("Cleaned:\n", text)
print("#######################################")


def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
print("Original:", text)
text = remove_punctuation(text)
print("Cleaned:\n", text)
print("#######################################")


def remove_numbers(text):
    return re.sub(r'\d+', '', text)
print("Original:", text)
text = remove_numbers(text)
print("Cleaned:\n", text)
print("#######################################")


def remove_special_apostrophes(text):
    return re.sub('[’]+', '', text)
print("Original:", text)
text = remove_special_apostrophes(text)
print("Cleaned:\n", text)
print("#######################################")


def remove_curly_quotes(text):
    return re.sub('[“”]', '', text)
print("Original:", text)
text = remove_curly_quotes(text)
print("Cleaned:\n", text)
print("#######################################")


```

```python
# optional

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english')) # Define stopwords set

# Removing stop words
def remove_stop_words(text):
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    cleaned_text = ' '.join(tokens)
    return cleaned_text
print("Original:", text)
text = remove_stop_words(text)
print("Cleaned:\n", text)
print("#######################################")
```

```python
# Or remove emojis

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U0001F1F2-\U0001F1F4"  # Macau flag
        u"\U0001F1E6-\U0001F1FF"  # flags
        u"\U0001F600-\U0001F64F"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U0001F1F2"
        u"\U0001F1F4"
        u"\U0001F620"
        u"\u200d"
        u"\u2640-\u2642"
        "]+", flags=re.UNICODE)

def remove_emojis(text):
    # return emoji.demojize(text, delimiters=(" ", " "))  # Optional: remove colons for cleaner text
     return emoji_pattern.sub(r'', text)
    
sample_text = "I'm so happy 😄! Let's grab some pizza 🍕 and celebrate 🎉.She loves traveling ✈️ and pizza 🍕!"
print("sample_text:", sample_text)
sample_text = remove_emojis(sample_text)
print("Cleaned:\n", sample_text)
print("#######################################")
```
