# Web scraping: NOTE YOU SHOULD TAKE PERMISSION FIRST

```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Natural_language_processing'

# 1. Add headers to avoid being blocked
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    response = requests.get(url, headers=headers, timeout=10)
    
    # Check if the request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract plain text from paragraphs
        paragraphs = [p.text for p in soup.find_all('p')]
        web_text = "\n".join(paragraphs)

        if web_text.strip():
            print(web_text[:500])
        else:
            print("No paragraph content found.")
    else:
        print(f"Failed to retrieve page. Status code: {response.status_code}")

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
import PyPDF2 # pip install PyPDF2 and restart kernel

with open('step1_universe.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    pdf_text = ''
    for page in reader.pages:
        pdf_text += page.extract_text()

print(len(pdf_text))
print(pdf_text[:500])
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

# link: https://www.youtube.com/watch?v=5ZFTY_eJR9k
video_id = "5ZFTY_eJR9k"

# link: https://www.youtube.com/watch?v=6g3iKLK1R8c
# video_id = "6g3iKLK1R8c"

try:
    ytt_api = YouTubeTranscriptApi()

    transcript = ytt_api.fetch(video_id)

    full_text = " ".join(snippet.text for snippet in transcript)

    print(full_text)

except Exception as e:
    print(e)
```
