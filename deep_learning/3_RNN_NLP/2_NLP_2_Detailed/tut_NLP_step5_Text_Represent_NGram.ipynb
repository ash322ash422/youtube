```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

# Download tokenizers if not already
nltk.download('punkt')

# Sample text
text = """
Softly they twinkle in velvet night,
Calming the soul with gentle light.
A breeze of peace floats through the air,
As starlit skies erase all care,
And hush the world in quiet delight.
"""

# Tokenize text into words
tokens = word_tokenize(text)

# Generate bigrams (2-grams)
bigrams = list(ngrams(tokens, 2))

# Generate trigrams (3-grams)
trigrams = list(ngrams(tokens, 3))

# Generate quadgrams (4-grams)
quadgrams = list(ngrams(tokens, 4))

print("Tokens:", tokens)
print("\nBigrams:")
for bg in bigrams:
    print(bg)

print("\nTrigrams:")
for tg in trigrams:
    print(tg)

print("\nQuadgrams:")
for qg in quadgrams:
    print(qg)
```
