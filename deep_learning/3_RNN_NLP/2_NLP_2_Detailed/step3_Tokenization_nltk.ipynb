```python
# The Punkt tokenizer is a data-driven sentence tokenizer that comes with NLTK.
# It is trained on large corpus of text to identify sentence boundaries.

import nltk.data
text = "hey , check new iphone launch? internet! think theyve finally added usb-c. \
yesterday, visited see specs, honestly, looks iphone . also, lol like, basically \
glorified iphone s.  face_with_tears_of_joy  mean, get wrong—it better camera, \
sure, $? thats much. btw, watched netflix show tech diaries? got episodes season . \
theres funny scene robot says: humans obsolete. classic. anyway, im thinking switching android. \
new pixel seems promising! lets catch soon—maybe saturday burgur palace?  hamburger \
oh found article (in verge) claiming battery life improved. talk soon! cheers, alex"

# Loading PunktSentenceTokenizer using English pickle file
tokenizer = nltk.data.load('tokenizers/punkt/PY3/english.pickle')
tokenizer.tokenize(text)
```

```python
# !pip install spacy
# !python -m spacy download en_core_web_sm

import spacy 

text = "hey , check new iphone launch? internet! think theyve finally added usb-c. \
yesterday, visited see specs, honestly, looks iphone . also, lol like, basically \
glorified iphone s.  face_with_tears_of_joy  mean, get wrong—it better camera, \
sure, $? thats much. btw, watched netflix show tech diaries? got episodes season . \
theres funny scene robot says: humans obsolete. classic. anyway, im thinking switching android. \
new pixel seems promising! lets catch soon—maybe saturday burgur palace?  hamburger \
oh found article (in verge) claiming battery life improved. talk soon! cheers, alex"

# Use larger or more accurate model (like en_core_web_md or en_core_web_trf) for deeper NLP tasks
nlp = spacy.load("en_core_web_sm")
# doc = nlp("Hello there! How are you? I hope you're doing well. Let's meet tomorrow.")
doc = nlp(text)
sentences = [sent.text for sent in doc.sents]
print(sentences)
```

```python
from nltk.tokenize import word_tokenize
 
text = "Hello!! Welcome to the class. Are we going to tokenize this sentence?? It's fine ;)"
print(word_tokenize(text))
```

```python
# Another word tokeizer
from nltk.tokenize import TreebankWordTokenizer

text = "Hello!! Welcome to the class. Are we going to tokeize this sentence?? It's fine ;)"

tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(text))
```

```python
# The WordPunctTokenizer is one of the NLTK tokenizers that splits words
# based on punctuation boundaries. Each punctuation mark is treated as a separate token.

from nltk.tokenize import WordPunctTokenizer

text = "Hello!! Welcome to the class. Are we going to tokenize this sentence?? It's fine ;)"
tokenizer = WordPunctTokenizer()
tokenizer.tokenize(text)
```
