## Synonym

```python
# Synonym
import nlpaug.augmenter.word as naw
import random
import numpy as np

# Set seed for reproducibility
random.seed(42)
np.random.seed(42)

# Input sentence
test_sentence = 'The food was good. The food was delicious. Probably because Rahim was hungry. Also, Rahim likes spicy food'

# Create synonym augmenter
aug = naw.SynonymAug(
    aug_src='wordnet',
    name='Synonym_Aug',
    aug_min=1,
    aug_max=10,
    aug_p=0.3,
    lang='eng',
    force_reload=False,
    verbose=0
)

# Generate exactly 5 augmented outputs
test_sentence_aug_list = aug.augment(test_sentence, n=5)

# Print results
print("Original:", test_sentence)
print("\nAugmented Outputs:")
for i, aug_sent in enumerate(test_sentence_aug_list, 1):
    print(f"{i}: {aug_sent}")
```

## Adding noise

```python
# keyboard proximity error


import nlpaug.augmenter.char as nac
import random
import numpy as np

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

test_sentence = 'I went Shopping Today, and bought bunch of bananas.'

# Force exactly 3 character-level augmentations
aug = nac.KeyboardAug(
    name='Keyboard_Aug',
    aug_char_min=1,
    aug_char_max=1,
    aug_char_p=0.4,          # Apply augmentation with high certainty
    aug_word_p=0.3,
    aug_word_min=1,
    aug_word_max=10,
    stopwords=None,
    tokenizer=None,
    reverse_tokenizer=None,
    include_special_char=True,
    include_numeric=True,
    include_upper_case=True,
    lang='en',
    verbose=0,
    stopwords_regex=None,
    model_path=None,
    min_char=4
)

# Perform augmentation
test_sentence_aug = aug.augment(test_sentence)

# Output
print("Original:", test_sentence)
print("Augmented:", test_sentence_aug)
```

```python
# spelling error

import nlpaug.augmenter.word as naw
import random
import numpy as np

# Set seed for reproducibility
random.seed(42)
np.random.seed(42)

# Input sentence
test_sentence = 'The food lacked taste.'

aug = naw.SpellingAug(dict_path=None, name='Spelling_Aug', aug_min=1,
                      aug_max=1, aug_p=0.3, stopwords=None, 
                      tokenizer=None, reverse_tokenizer=None, include_reverse=True, stopwords_regex=None, verbose=0)
 
test_sentence_aug = aug.augment(test_sentence, n=3)
print(test_sentence)
print(test_sentence_aug)
```

```python
# Random insertion, deletion, or swap

import nlpaug.augmenter.char as nac

text = "This is another example to add noise."
aug = nac.RandomCharAug(action="insert", aug_char_p=0.001)
augmented_insert = aug.augment(text)

aug = nac.RandomCharAug(action="delete", aug_char_p=0.02)
augmented_delete = aug.augment(text)

aug = nac.RandomCharAug(action="swap", aug_char_p=0.02)
augmented_swap = aug.augment(text)

print("Original:", text)
print("Insert Noise:", augmented_insert)
print("Delete Noise:", augmented_delete)
print("Swap Noise:", augmented_swap)
```
