```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus (list of documents)

corpus = [
    "I love the trees, love the whispers.",
    "Forest trees sing, soothing trees melodies.",
    "Morning dawn, I love the trails.",
    ]

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus to BoW representation
X = vectorizer.fit_transform(corpus)

# Get feature (vocabulary) names
print("Vocabulary:")
print(vectorizer.get_feature_names_out())

# Show Bag of Words matrix
print("\nBoW Matrix (Document-Term Matrix):")
print(X.toarray())
```
