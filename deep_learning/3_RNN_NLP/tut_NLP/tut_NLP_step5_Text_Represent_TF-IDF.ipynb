```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample corpus (list of documents)
corpus = [
    "I love the trees, the whispers.",
    "Forest trees sing, soothing melodies.",
    "Morning dawn, I love the trails.",
    "Enchanting winds, the trees dance.",
    "I explore, explore the secret paths."
]

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus to TF-IDF representation
X = vectorizer.fit_transform(corpus)

# Get feature (vocabulary) names
print("Vocabulary:")
print(vectorizer.get_feature_names_out())

# Show TF-IDF matrix (Document-Term Matrix)
print("\nTF-IDF Matrix (Document-Term Matrix):")
print(X.toarray())
```

```python
# Experimenting with min_df, max_df, and stop_words in TfidfVectorizer

from sklearn.feature_extraction.text import TfidfVectorizer

# Sample corpus (list of documents)
corpus = [
    "I love the trees, the whispers.",
    "Forest trees sing, soothing melodies.",
    "Morning dawn, I explore the trails.",
    "love enchanting winds, the trees dance.",
    "I explore the secret paths."
]

# Initialize the TfidfVectorizer with custom parameters
vectorizer = TfidfVectorizer(
    min_df=2,                  # Include words that appear in at least 2 documents
    max_df=0.8,                # Exclude words that appear in more than 80% of documents
    stop_words='english'       # Remove common English stop words
)

# Fit and transform the corpus to TF-IDF representation
X = vectorizer.fit_transform(corpus)

# Get feature (vocabulary) names
print("Vocabulary (After applying min_df, max_df, and stop_words):")
print(vectorizer.get_feature_names_out())

# Show TF-IDF matrix (Document-Term Matrix)
print("\nTF-IDF Matrix (Document-Term Matrix):")
print(X.toarray())
```
