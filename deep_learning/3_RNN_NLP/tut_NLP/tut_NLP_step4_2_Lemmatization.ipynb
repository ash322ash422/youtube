```python
# using spacy
import spacy
nlp = spacy.load("en_core_web_sm") # Load a pre-trained spaCy model
# text = "I am running to the store because the store has better apples"
# text = "The human brain is capable of complex language processing, but \
# decoding thoughts directly from brain activity has long been a challenge. \
# Previous research has attempted this by using classification models "

text = "The children were running and playing in the fields while \
their parents watched from a distance. They had brought snacks, blankets, \
and games to enjoy the sunny afternoon. As the sun began setting, the families \
started packing their belongings, smiling and chatting about the experiences of the day. \
Some kids had taken photos, others had been drawing or painting nature scenes. It was \
a joyful gathering filled with laughter, shared stories, and lasting memories. "
doc = nlp(text)
lemmas = [token.lemma_ for token in doc]
print(f"lemmas: {lemmas}")

lemmatized_text = " ".join(lemmas)
print(f"lemmatized_text: {lemmatized_text}")
```
