```python
# Style1: Using regex. This allows you to customize the contractions. Also it does not use external
#  library
import re

# Dictionary of common contractions
contractions_dict = {
    "I'm": "I am",
    "you're": "you are",
    "he's": "he is",
    "she's": "she is",
    "it's": "it is",
    "we're": "we are",
    "they're": "they are",
    "I've": "I have",
    "you've": "you have",
    "we've": "we have",
    "they've": "they have",
    "can't": "cannot",
    "won't": "will not",
    "didn't": "did not",
    "don't": "do not",
    "doesn't": "does not",
    "haven't": "have not",
    "hadn't": "had not",
    "hasn't": "has not",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "wouldn't": "would not",
    "shouldn't": "should not",
    "couldn't": "could not",
    "mustn't": "must not",
    "mightn't": "might not",
    "needn't": "need not",
    "I'll": "I will",
    "you'll": "you will",
    "he'll": "he will",
    "she'll": "she will",
    "we'll": "we will",
    "they'll": "they will",
    "I'd": "I would",
    "you'd": "you would",
    "he'd": "he would",
    "she'd": "she would",
    "we'd": "we would",
    "they'd": "they would",
    "there's": "there is",
    "that's": "that is",
    "what's": "what is",
    "who's": "who is",
    "let's": "let us"
}

# Precompile regex pattern for efficiency
contractions_re = re.compile('(%s)' % '|'.join(map(re.escape, contractions_dict.keys())), flags=re.IGNORECASE)

# Function to expand contractions
def expand_contractions(text):
    def replace(match):
        contraction = match.group(0)
        expanded = contractions_dict.get(contraction)
        if not expanded:
            expanded = contractions_dict.get(contraction.lower())
        # Preserve case
        if contraction[0].isupper():
            expanded = expanded[0].upper() + expanded[1:]
        return expanded

    return contractions_re.sub(replace, text)

# Sample text
text = """I'm sure you've seen it. He didn't know they weren't coming, and that's why he wasn't ready."""

# Expand contractions
expanded_text = expand_contractions(text)

print("Original Text:")
print(text)
print("\nExpanded Text:")
print(expanded_text)
```
