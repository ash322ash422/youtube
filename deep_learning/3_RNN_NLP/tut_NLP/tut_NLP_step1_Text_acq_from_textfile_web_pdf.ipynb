```python
# from  website: NOTE YOU SHOULD TAKE PERMISSION FIRST
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Natural_language_processing'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract plain text from paragraphs
paragraphs = [p.text for p in soup.find_all('p')]
web_text = "\n".join(paragraphs)

print(web_text[:500])
```

```python
# From pdf
import PyPDF2 # pip install PyPDF2

with open('step1_universe.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    pdf_text = ''
    for page in reader.pages:
        pdf_text += page.extract_text()

print(len(pdf_text))
print(pdf_text[:500])
```

```python
# from Youtube
from youtube_transcript_api import YouTubeTranscriptApi # pip install youtube-transcript-api

# https://www.youtube.com/watch?v=88g9XiZava4&t=84s
video_id = "88g9XiZava4&t=84s"
transcript = YouTubeTranscriptApi.get_transcript(video_id)

# Combine all parts of the transcript
full_text = " ".join([segment['text'] for segment in transcript])
print(full_text)
```
