# TF-IDF implementation

```python
# documents = [
#     "Delta offers great service and friendly staff",
#     "United provides great in-flight entertainment and service",
#     "Delta and United both offer good service. Delta provides comfort"
# ]
documents = [
"Delta provides great service",
"United offers great food",
"Great airlines offer comfort"
]
```

```python
# step1) Preprocessing (lowercase + tokenize)

import re
from collections import Counter
import numpy as np
import math

def preprocess(doc):
    doc = doc.lower()
    doc = re.sub(r"[^a-z\s]", "", doc)
    return doc.split()

tokenized_docs = [preprocess(doc) for doc in documents]
print(tokenized_docs)
```

```python
# Step 2) Compute Term Frequency (TF) for each document

# Build vocabulary
vocab = sorted(set(word for doc in tokenized_docs for word in doc))
print(f"vocab: {vocab}")

def compute_tf(doc_tokens):
    word_count = Counter(doc_tokens)
    total_words = len(doc_tokens)
    return {word: word_count[word] / total_words for word in vocab}

tf_scores = [compute_tf(doc) for doc in tokenized_docs]

print("TF scores for each documents:")
for tf_score in tf_scores:
    print(tf_score)
```

```python
# Step 3) Compute Inverse Document Frequency (IDF)
def compute_idf(docs):
    N = len(docs)
    idf_scores = {}
    for word in vocab:
        containing_docs = sum(1 for doc in docs if word in doc)
        idf_scores[word] = math.log(N / (1 + containing_docs)) + 1  # add-1 smoothing, and log is to the base e
    return idf_scores

idf_scores = compute_idf(tokenized_docs)
print(idf_scores)
```

```python
# Step 4: Compute TF-IDF for each document
tfidf_scores = []
for doc_tf in tf_scores:
    tfidf = {word: doc_tf[word] * idf_scores[word] for word in vocab}
    tfidf_scores.append(tfidf)

for tfidf_score in tfidf_scores:
    print(tfidf_score)
```

```python
# Step 5: View TF-IDF vector (rounded)
import pandas as pd

df = pd.DataFrame(tfidf_scores)
print(df.round(3))
```
