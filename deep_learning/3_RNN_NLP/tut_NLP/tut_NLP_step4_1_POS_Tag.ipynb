```python
# using spacy

import spacy

text="""The human brain is capable of complex language processing, but decoding thoughts directly from 
brain activity has long been a challenge. Previous research has attempted this by using classification models 
that match brain activity to predefined language options
"""

nlp = spacy.load("en_core_web_sm")
# doc = nlp("The quick brown fox jumps over the lazy dog.")
doc = nlp(text)
POS = []
for token in doc:
    POS.append((token.text,token.pos_))
    # print(f"{token.text}: {token.pos_}")

print(POS)
```
