```python
# PorterStemmer
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download necessary resources
nltk.download('punkt')

# Initialize the stemmer
stemmer = PorterStemmer()

# Sample paragraph with many stemmable words
text = """
He was running towards the finishing line, hoping to be declared the winner of the competition.
However, the organizers were still finalizing the arrangements and were not prepared for his arrival.
Many spectators were cheering, clapping, and celebrating the achievements of their favorite participants.
"""

# Tokenize the paragraph into words
words = word_tokenize(text)

# Apply stemming
stemmed_words = [stemmer.stem(word) for word in words]

# Display original and stemmed words
for original, stemmed in zip(words, stemmed_words):
    print(f"{original:15} --> {stemmed}")
```

```python
# SnowballStemmer: supports multiple languages 
import nltk
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize

# Download required resources
nltk.download('punkt')

# Initialize the Snowball stemmer for English
stemmer = SnowballStemmer("english") # supports multiple languages 

# Sample paragraph
text = """
He was running towards the finishing line, hoping to be declared the winner of the competition.
However, the organizers were still finalizing the arrangements and were not prepared for his arrival.
Many spectators were cheering, clapping, and celebrating the achievements of their favorite participants.
"""

# Tokenize text into words
words = word_tokenize(text)

# Apply stemming
stemmed_words = [stemmer.stem(word) for word in words]

# Display original vs. stemmed words
for original, stemmed in zip(words, stemmed_words):
    print(f"{original:15} --> {stemmed}")
```
