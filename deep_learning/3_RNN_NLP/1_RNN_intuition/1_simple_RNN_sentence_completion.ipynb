# Step 1: Tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)
```

```python
print("\nWord to Index Mapping:\n")
print(tokenizer.word_index)
```

# Step 2: Create Input-Output Pairs

```python
input_sequences = []

for seq in sequences:
    for i in range(1, len(seq)):
        input_sequences.append(seq[:i+1])

print("input sequences:", input_sequences)
print("------------------------------------")
for s in input_sequences:
    print(s)
```

```python
[2, 1, 3, 4, 8]
i love going to movies
```

# Step 3: Padding

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_len = max([len(seq) for seq in input_sequences])

input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

print(input_sequences)
```

# Step 4: Split X and y

```python
import numpy as np

X = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

```python
X
```

```python
y
```

# Step 5: Build RNN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

vocab_size = len(tokenizer.word_index) + 1
embedding =  Embedding(vocab_size,
              # embedding dimension. I want you to represent each word with vector of 10 numbers
              10, 
              input_length=max_len-1
)
print("embedding (Before):", embedding.get_weights())

model = Sequential([
    embedding,
    SimpleRNN(50), # 50 = number of neurons (hidden units)
    Dense(vocab_size, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy']
)

model.summary()
```

# Step 6: Train Model

```python
model.fit(X, y, epochs=100, verbose=1)
```

```python
# Lets look at the embedding 

embedding_weights = model.layers[0].get_weights()[0]
print("embedding_weights:\n", embedding_weights)
print("-------------------------------")

word_index = tokenizer.word_index

word = "love"
index = word_index[word]

print(f"{word} -> {embedding_weights[index]}")
```

# Step 7: Prediction Function

```python
import numpy as np

def predict_next_word(text):
    seq = tokenizer.texts_to_sequences([text])[0]
    seq = pad_sequences([seq], maxlen=max_len-1, padding='pre')

    predict_probability = model.predict(seq, verbose=0)[0]
    # print("predict_probability:", predict_probability)

    # (OPTIONAL): print next word and its probability in sorted manner
    index_to_word = {index: word for word, index in tokenizer.word_index.items()}
    sorted_indices = np.argsort(predict_probability)[::-1]
    print("\nWord Probabilities:\n")
    for i in sorted_indices:
       word = index_to_word.get(i, "PAD/UNK")
       prob = predict_probability[i]
       print(f"{word} → {prob:.4f}")   
    # (END OPTIONAL) 
        
    predicted = np.argmax(predict_probability, axis=-1)
    
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            return word
```

```python
next_word = predict_next_word("i love") 
print("--------------")
print("next_word:", next_word)
```
