```python
#Extract all phone numbers from a string
# "The phone numbers are +91-9876543210, 9876543210, and (+91) 9876543210."

text = "The phone numbers are +91-9876543210, 9876543210, and (+91) 9876543210."

pattern = r"[+]+\d{1,3}[-\s]?\d{10}|\(?\d{3}\)?[-\s]?\d{3}[-\s]?\d{4}"

output = re.findall(pattern, text)

print(output)
```

```python
text = "The phone numbers are +91-9876543210, 9876543210, and (+91) 9876543210."

pattern = r"(?:\+91[-]?|\(\+91\)\s?)?\d{10}"

output = re.findall(pattern, text)

print(output)
```

```python
# re.search()

text = "Python is powerful."

pattern = r"powerful"

match = re.search(pattern, text)
print(match)
```

```python
if match:
    print("Found:", match)
    print("\n")
    print("Found: ", match.group())
else:
    print("Not found")
```

```python
# re.search()

text = "Python is powerful is more powerful."

pattern = r"powerful"

match = re.search(pattern, text)

if match:
    print("Found:", match)
    print("\n")
    print("Found: ", match.group())
else:
    print("Not found")
```

```python
# re.search()

text = "Python is powerful is more powerful."

pattern = r"powerful"

match = re.search(pattern, text)

if match:
    print("Found:", match)
    print("\n")
    print("Found: ", match.group())
    print("Start index:", match.start())
    print("End index:", match.end())
else:
    print("Not found")
```

```python
text = "My number is 12345."

pattern = r"\d+"
match = re.search(pattern, text)

print(match.group())
```

```python
text = "My number is 12345-678910."

pattern = r"\d+-\d+"
match = re.search(pattern, text)

print(match.group())
```

```python
text = "Date: 2025-06-10"

pattern = r"(\d{4})-(\d{2})-(\d{2})"

match = re.search(pattern, text)

if match:
    print("Year:", match.group(1))
    print("Month:", match.group(2))
    print("Day:", match.group(3))
```

```python
# re.sub()

text = "I love cats. Cats are cute."
new_text = re.sub(r"cats", "dogs", text)
print(new_text)
```

```python
text = "I love cats. Cats are cute."
new_text = re.sub(r"cats", "dogs", text, flags=re.IGNORECASE)
print(new_text)
```

```python
# Remove extra spaces from a string
text = "Python    is    great!"
cleaned = re.sub(r"\s+", " ", text)
print(cleaned)
```

```python
# Masking phone numbers in a string
text = "Call me at 9876543210 or 9123456789."
masked = re.sub(r"\d{6}(\d{4})", r"******\1", text)
print(masked)
```

```python
# Remove HTML tags from a string
html = "<div>Hello</div><p>World</p>"
text = re.sub(r"<.*?>", "", html)
print(text)
```

```python
# Replace first two occurrences of "two" with "2"
text = "one two three two four two"
result = re.sub(r"two", "2", text, count=2)
print(result)
```

```python
# re.compile() - Compile a pattern into a Pattern object - Compiles a regex pattern into a reusable object.

pattern = re.compile(r"\d+")
pattern.findall("123 456")
```

```python
# email pattern with re.compile()

pattern = re.compile(r"[\w\.-]+@[\w\.-]+")
text = "Please contact me at anantv0@gmail.com and abc-34d@gmail.com"
output = pattern.findall(text)
print(output)

```
