````python
import gensim
from gensim.models import Word2Vec
````

## Sample Corpus (Tokenized Text)

 Gensim expects **list of list of words**


````python
corpus = [
    ["i", "love", "machine", "learning"],
    ["machine", "learning", "is", "fun"],
    ["deep", "learning", "is", "powerful"],
    ["i", "enjoy", "learning", "new", "things"],
    ["artificial", "intelligence", "is", "the", "future"]
]
````

## Train Word2Vec Model

````python
# Convert words to vector

w2v_model = Word2Vec(
    sentences=corpus,
    vector_size=50,   # size of word vectors
    window=3,         # context window
    min_count=1,      # include all words
    workers=2
)
````

## Word Embeddings (Core Idea)

````python
vector = w2v_model.wv['learning']
print(vector)
````

##  Vocabulary

````python
print(w2v_model.wv.index_to_key)
````

## Similar Words (VERY IMPORTANT DEMO)

````python
w2v_model.wv.most_similar('learning')
````

##  Similarity Between Words

````python
sim = w2v_model.wv.similarity('machine', 'learning')
print(sim)
````

## Words Not in Vocabulary

````python
print("ai" in w2v_model.wv)
````

## Save & Load Model

````python
w2v_model.save("word2vec.model")

# Load later
model_loaded = Word2Vec.load("word2vec.model")
````

````python
model_loaded.wv.most_similar('learning')
````

## Connect to RNN (VERY IMPORTANT)
Convert words → vectors → feed into LSTM

````python
sentence = ["i", "love", "learning"]

vectors = [w2v_model.wv[word] for word in sentence]

print(vectors)
````

````python
Now this becomes input to LSTM
````

### Pipeline:

```
Text → Tokenization → Word2Vec → Vectors → LSTM
```
