# LSTM

## Step 1: Import Libraries

````python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
````

````python
# 1) Small number of sentences

sentences = [
    "I love machine learning",
    "Deep learning is powerful",
    "AI is the future"
]
````

## Step 3: Tokenization

````python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)
````

## Step 4: Create Input-Output Pairs

````python
X = []
y = []

for seq in sequences:
    for i in range(1, len(seq)):
        X.append(seq[:i])
        y.append(seq[i])

X = pad_sequences(X)
y = np.array(y)
````

````python
print(X[:3])
print(y[:3])
````

## Step 5: Build LSTM Model

````python
vocab_size = len(tokenizer.word_index) + 1 # +1 is for padding (index 0)
print(vocab_size)
````

````python
vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size, 
              output_dim=10, 
              input_length=X.shape[1]),
    LSTM(64), # GRU(64), # For GRU: Replace LSTM with GRU
    Dense(vocab_size, activation='softmax')
])
````

## Step 6: Compile Model

````python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
````

## Step 7: Train

````python
model.fit(X, y, epochs=100)
````

# Now we use the model: Predict next word

The flow

```
Text → Tokenizer → Sequence → Padding → Model → Prediction → Convert back to word


## Step 1: Input sentence

````python
# input_text = "I love"
````

````python
# For longer sentences use following:

input_text = "I love"
# input_text = "Natural language"

````

## Step 2: Convert to sequence

````python
seq = tokenizer.texts_to_sequences([input_text])
print(seq)
````

## Step 3: Pad sequence (VERY IMPORTANT)

Model expects same length as training data

````python
seq = pad_sequences(seq, maxlen=X.shape[1])
print(seq)
````

## Step 4: Predict

````python
pred = model.predict(seq)
print(pred)
````

## Step 5: Get predicted word index

````python
predicted_index = np.argmax(pred)
print(predicted_index)
````

## Step 6: Convert index → word

````python
for word, index in tokenizer.word_index.items():
    if index == predicted_index:
        print(word)
````

# Generate full Sentence

````python
def generate_text(seed_text, next_words):
    for _ in range(next_words):
        seq = tokenizer.texts_to_sequences([seed_text])
        seq = pad_sequences(seq, maxlen=X.shape[1])

        pred = model.predict(seq)
        predicted_index = np.argmax(pred)

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break

        seed_text += " " + output_word

    return seed_text

print(generate_text("I love", 3))
````
