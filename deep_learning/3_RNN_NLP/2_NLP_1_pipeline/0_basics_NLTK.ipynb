## Sentence Tokenization

````python
from nltk.tokenize import sent_tokenize

text = "I love learning about Artificial Intelligence. It is transforming the world! It is number 6 in my favorite things"

sentences = sent_tokenize(text)

print(sentences)
````

## 4. Word Tokenization

````python
from nltk.tokenize import word_tokenize

text = "I love learning about Artificial Intelligence. It is transforming the world! It is number 6 in my favorite things"

words = word_tokenize(text)

print(words)
````

## Lowercase + Remove Punctuation

````python
import string

words_clean = [word.lower() for word in words if word.isalpha()]
print(words_clean)
````

## Remove Stopwords

````python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [w for w in words_clean if w not in stop_words]

print(filtered_words)
````

Removes: `is, the, about, etc.`

##  Stemming

```


# Why stemming: 
- Reduces vocabulary size 
- model performances - I learned cooking. I am learning cooking   ..after stemming,  model sees learned and learning as same

````python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

# A simple list of 20 words that share common stems
word_list = [
    "connection", "connected", "connecting", "connects",
    "programming", "programs", "programmer", "programmers",
    "compute", "computer", "computing", "computation",
    "argue", "arguing", "argued", "argues",
    "study", "studying", "studied", "studies"
]

# Print the results
print(f"{'Original Word':<15} | {'Stemmed Output'}")
print("-" * 35)

for word in word_list:
    print(f"{word:<15} | {stemmer.stem(word)}")
````

````python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(w) for w in filtered_words]

print("original:", filtered_words)
print("stemmed :", stemmed_words)
````
