```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

# model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
model_name ="dslim/bert-base-NER"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

ner = pipeline("ner", model=model, tokenizer=tokenizer)

text = "Elon Musk founded Tesla in the USA"
result = ner(text)
print(result)
```

```python
for r in result:
  print(r)

# The model dslim/bert-base-NER is based on BERT, which uses a
# WordPiece tokenizer. This tokenizer breaks words into smaller subword
# units when the full word is not stored as a single token in the model
#  vocabulary.
```

```python
texts = [
    "Elon Musk founded Tesla in the USA",
    "Apple is a company",
    "I ate an apple" # This is empty because it was trained on a specific type of dataset
]

for t in texts:
    print("\n", t)
    print(ner(t))
```
