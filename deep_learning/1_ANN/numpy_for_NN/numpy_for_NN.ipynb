# Performance on matrix multiplication

````python
A = np.array([
    [2, 4, 1 ],
    [7, 2, 5 ]
])

B = np.array([
    [4, 5, 2],
    [6, 8, 5],
    [2, 2, 1]
])

result = A @ B   # or A @ B, np.matmul(A, B)

print(result)
````

# Vectorized operation

````python
arr1 = np.array([3, 2, 7, 9, 0])
arr2 = np.array([4, 1, 6, 8, 4])

result = arr1 * arr2  # vectorized operation

print(result)
````

# A simple NN

````python
x = np.array([1, 2, 3])
w = np.array([0.2, 0.8, -0.5])
b = 2

z = w @ x  +  b
print(z)
````

````python
# Activation function

output =  np.maximum(0, z)
print(output)
````

# Multiple Neuron Layer

````python
# Input (3 features)
x = np.array([1.0, 2.0, 3.0])

# 3 neurons → each has 3 weights
W = np.array([
    [0.20, 0.80, -0.50],
    [0.50, -0.91, 0.26],
    [-0.26, -0.27, 0.17]
])

# Bias for each neuron
b = np.array([2, 3, 0.5])

# Layer output
Z = W @ x  +  b

print("Layer Output:\n", Z)
````

# Batch Input

````python
# 2 samples
X = np.array([
    [1, 2, 3],
    [2, 5, -1]
])

Z = X @ (W.T)  +  b
print(Z)
````

# Mini Assignment

Build this:

Input:
```
X = [[1,2,3], [4,5,6]]
```

## Task:

Create:
- 4 neurons
- random weights
- random bias
- Output final matrix

````python
# Solution
import numpy as np

# Input (2 samples, 3 features)
X = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Step 1: Initialize random weights (4 neurons, 3 inputs each)
W = np.random.randn(4, 3)

# Step 2: Initialize random bias (1 per neuron)
b = np.random.randn(4)

# Step 3: Compute output
Z = X @ (W.T) + b

print("Weights:\n", W)
print("\nBias:\n", b)
print("\nOutput:\n", Z)
````
