# CPU vs GPU
- I ran following on Google Colab
- I took 2 matrix A and B, each of shape 4000 x 4000
- I multiplied them on CPU first. It took 3.88 sec
- I multiplied them on GPU. It took 0.04 sec


# CPU

```python
import numpy as np
import time

A = np.random.rand(4000,4000)
B = np.random.rand(4000,4000)

start = time.time()

C = np.dot(A,B)

end = time.time()

print("CPU time:", end-start) # 3.88
```

# GPU

```python
import torch
import time

device = "cuda"

A = torch.randn(4000,4000).to(device)
B = torch.randn(4000,4000).to(device)

start = time.time()

C = torch.matmul(A,B)

torch.cuda.synchronize()

end = time.time()

print("GPU time:", end-start) # 0.04
```
