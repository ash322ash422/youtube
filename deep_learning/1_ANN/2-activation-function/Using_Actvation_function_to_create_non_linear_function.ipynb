# Using Linear ReLUs Create a Non-Linear Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input data
x = np.linspace(0, 5, 500)

# Define the basic ReLU function
def relu(val):
    return np.maximum(0, val)

# 1. Calculate individual neuron outputs
n1 = relu(1 * x - 1)
n2 = relu(2 * x - 4)
n3 = relu(-1 * x + 3)

# 2. Combine them in the final output layer
combined_output = n1 - n2 + n3

# 3. Plot the results
plt.figure(figsize=(8, 5))
plt.plot(x, n1, '--', label='Neuron 1: relu(x - 1)', alpha=0.6)
plt.plot(x, n2, '--', label='Neuron 2: relu(2x - 4)', alpha=0.6)
plt.plot(x, n3, '--', label='Neuron 3: relu(-x + 3)', alpha=0.6)
plt.plot(x, combined_output, color='black', linewidth=3,
         label='Combined Non-Linear Output'
)

plt.title('How 3 Linear ReLUs Create a Non-Linear Function')
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Simulating y = x² Using ReLU Activation

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate inputs
x = np.linspace(-3, 3, 600)
actual_y = x**2

# Define ReLU
def relu(val):
    return np.maximum(0, val)

# Right side approximation (Positive x)
r1 = 1 * relu(x)
r2 = 2 * relu(x - 1)
r3 = 2 * relu(x - 2)

# Left side approximation (Negative x)
l1 = 1 * relu(-x)
l2 = 2 * relu(-x - 1)
l3 = 2 * relu(-x - 2)

# Combine all neurons (Bias offset accounts for the base)
simulated_y = r1 + r2 + r3 + l1 + l2 + l3

# Plotting
plt.figure(figsize=(7, 4.5))
plt.plot(x, actual_y, 'r--', label='Actual: y = x²', linewidth=2)
plt.plot(x, simulated_y, 'b-', label='ReLU Simulation (6 Neurons)', linewidth=3)

plt.title('Simulating y = x² Using ReLU Activation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
