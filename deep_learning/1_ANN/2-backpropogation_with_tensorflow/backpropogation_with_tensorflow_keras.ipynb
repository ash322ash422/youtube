# Step-by-Step Implementation

Example: Single Neuron Training

# Step 1: Setup

````python
import tensorflow as tf

# Input and true output
X = tf.constant([[1.0, 2.0, 3.0]])
y_true = tf.constant([[10.0]])
````

# Step 2: Initialize Weights - Random

````python
W = tf.Variable([[0.5], [0.3], [-0.2]])
b = tf.Variable([0.1])
````

# Step 3: Forward + Backward Pass

Flow
- Forward pass: Compute prediction
- Loss: Measure error
- Backward: Use calculus to adjust weights to reduce loss

````python
learning_rate = 0.01

for epoch in range(10):
    with tf.GradientTape() as tape:
       
        # Forward pass
        z = tf.matmul(X, W) + b
        y_pred = tf.nn.relu(z) # Activation (ReLU)
        
        # Loss (MSE)
        loss = tf.reduce_mean((y_true - y_pred) ** 2)

    # Backward pass (gradients)
    grads = tape.gradient(loss, [W, b])

    # Update weights
    W.assign_sub(learning_rate * grads[0])
    b.assign_sub(learning_rate * grads[1])

    print(f"Epoch {epoch}: Loss = {loss.numpy()}")
````

````python
# print W and b
print("W:", W)
print("b:", b)
````

````python
# prediction
X = tf.constant([[1.0, 2.0, 3.0]])
z = tf.matmul(X, W) + b
y_pred = tf.nn.relu(z) # Activation (ReLU)
print(y_pred)
````

# Extend to Multi-Neuron Layer

````python
# Input (2 samples, 3 features)
X = tf.constant([
    [1.0, 2.0, 3.0],
    [2.0, 5.0, -1.0]
])

# True outputs (2 samples, 3 neurons)
y_true = tf.constant([
    [1.0, 0.0, 1.0],
    [0.0, 1.0, 0.0]
])

# Weights (3 inputs в†’ 3 neurons)
W = tf.Variable(tf.random.normal((3, 3)))
b = tf.Variable(tf.random.normal((3,)))

for epoch in range(10):

    with tf.GradientTape() as tape:
        z = tf.matmul(X, W) + b
        y_pred = tf.nn.relu(z) # activation
        
        loss = tf.reduce_mean((y_true - y_pred) ** 2)

    grads = tape.gradient(loss, [W, b])

    W.assign_sub(0.01 * grads[0])
    b.assign_sub(0.01 * grads[1])

    print(f"Epoch {epoch}: Loss = {loss.numpy()}")
````

# 2-Layer Neural Network (From Scratch with Training)
Architecture
```
Input (3 features)
   в†“
Hidden Layer (4 neurons, ReLU)
   в†“
Output Layer (1 neuron)
```


## Core Math

Forward pass:

Z1 = X * W1 + b1

A1 = max(0, Z1)

Z2 = A1 * W2 + b2

````python
import tensorflow as tf

# 1. Data
X = tf.constant([
    [1.0, 2.0, 3.0],
    [2.0, 5.0, -1.0],
    [-1.0, 2.0, 4.0]
])

y_true = tf.constant([
    [1.0],
    [0.0],
    [1.0]
])

# 2. Initialize Weights: # input в†’ hidden
W1 = tf.Variable(tf.random.normal((3, 4)))  
b1 = tf.Variable(tf.zeros((4,)))

# hidden в†’ output
W2 = tf.Variable(tf.random.normal((4, 1)))  
b2 = tf.Variable(tf.zeros((1,)))

learning_rate = 0.01

# 3. Training Loop
for epoch in range(50):

    with tf.GradientTape() as tape:
        # ---- Forward Pass ----
        Z1 = tf.matmul(X, W1) + b1
        A1 = tf.nn.relu(Z1)

        Z2 = tf.matmul(A1, W2) + b2
        y_pred = tf.nn.sigmoid(Z2)   # output activation

        # ---- Loss ----
        loss = tf.reduce_mean((y_true - y_pred) ** 2)

    # ---- Backprop ----
    grads = tape.gradient(loss, [W1, b1, W2, b2])

    # ---- Update ----
    W1.assign_sub(learning_rate * grads[0])
    b1.assign_sub(learning_rate * grads[1])
    W2.assign_sub(learning_rate * grads[2])
    b2.assign_sub(learning_rate * grads[3])

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.numpy()}")
````

````python
# 4. Final Predictions
X_sample = tf.constant([
    [1.0, 2.0, 3.0],
    [2.0, 5.0, -1.0],
    [-1.0, 2.0, 4.0]
])
Z1 = tf.matmul(X_sample, W1) + b1
A1 = tf.nn.relu(Z1)

Z2 = tf.matmul(A1, W2) + b2
y_pred = tf.nn.sigmoid(Z2)   

print("\nFinal Predictions:\n", y_pred.numpy())
````

# Redo above using Kera (5 lines of code)

````python
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(4, activation='relu', 
                       input_shape=(3,)),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y_true, epochs=50)
````

````python
X_sample = tf.constant([
    [1.0, 2.0, 3.0],
    [2.0, 5.0, -1.0],
    [-1.0, 2.0, 4.0]
])

predictions = model.predict(X_sample)
print("Predictions:\n", predictions)
````
