```python
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-5, 5, 200)

# Weight and bias
w = 1.2
b = -0.5

# Net input
n = w * x + b

# Activation functions
def purelin(n):
    return n

def hardlim(n):
    return np.where(n >= 0, 1, 0)

def hardlims(n):
    return np.where(n >= 0, 1, -1)

def tansig(n):
    return 2 / (1 + np.exp(-2*n)) - 1   # same as tanh

def logsig(n):
    return 1 / (1 + np.exp(-n))

# Outputs
a_purelin  = purelin(n)
a_hardlim  = hardlim(n)
a_hardlims = hardlims(n)
a_tansig   = tansig(n)
a_logsig   = logsig(n)

# Plotting
plt.figure(figsize=(10,10))

plt.subplot(3,2,1)
plt.plot(x, a_purelin, 'b', linewidth=2)
plt.title('purelin (Linear)')
plt.grid(True)

plt.subplot(3,2,2)
plt.step(x, a_hardlim, 'r', where='mid')
plt.title('hardlim (Binary 0/1)')
plt.grid(True)

plt.subplot(3,2,3)
plt.step(x, a_hardlims, 'g', where='mid')
plt.title('hardlims (Binary -1/1)')
plt.grid(True)

plt.subplot(3,2,4)
plt.plot(x, a_tansig, 'm', linewidth=2)
plt.title('tansig (Sigmoid -1 to 1)')
plt.grid(True)

plt.subplot(3,2,5)
plt.plot(x, a_logsig, 'c', linewidth=2)
plt.title('logsig (Sigmoid 0 to 1)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Activation functions
# --------------------------
def purelin(n):
    return n

def hardlim(n):
    return np.where(n >= 0, 1, 0)

# --------------------------
# Generate toy dataset
# --------------------------
np.random.seed(0)
# Class 0: points near -2
x_class0 = np.random.randn(20) - 2
y_class0 = np.zeros_like(x_class0)

# Class 1: points near +2
x_class1 = np.random.randn(20) + 2
y_class1 = np.ones_like(x_class1)

# Combine
X = np.concatenate([x_class0, x_class1])
Y = np.concatenate([y_class0, y_class1])

# --------------------------
# Neuron parameters
# --------------------------
w = 1.0
b = 0.0
n = w*X + b

# Predictions
y_purelin = purelin(n)
y_hardlim = hardlim(n)

# --------------------------
# Plot results
# --------------------------
plt.figure(figsize=(12,5))

# Purelin output (continuous)
plt.subplot(1,2,1)
plt.scatter(X, Y, c=Y, cmap='bwr', edgecolor='k', s=70, label='True class')
plt.plot(X, y_purelin, 'ko', label='Neuron output')
plt.axhline(0.5, color='gray', linestyle='--')
plt.title("Neuron with purelin (continuous)")
plt.xlabel("Input X"); plt.ylabel("Output")
plt.legend()

# Hardlim output (binary)
plt.subplot(1,2,2)
plt.scatter(X, Y, c=Y, cmap='bwr', edgecolor='k', s=70, label='True class')
plt.plot(X, y_hardlim, 'ko', label='Neuron output')
plt.title("Neuron with hardlim (classification)")
plt.xlabel("Input X"); plt.ylabel("Output")
plt.legend()

plt.tight_layout()
plt.show()
```
