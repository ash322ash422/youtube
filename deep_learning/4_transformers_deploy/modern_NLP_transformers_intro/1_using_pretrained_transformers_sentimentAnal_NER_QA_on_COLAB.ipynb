Task 1: Sentiment Analysis (Easiest Demo)

```python
# Took 3 minutes
from transformers import pipeline

# Load pretrained pipeline
classifier = pipeline("sentiment-analysis")

# Test
result = classifier("I love learning AI and NLP!")
print(result)

result = classifier("I hate learning AI and NLP!")
print(result)
```

# Task 2: Named Entity Recognition (NER)

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

text = "John works at Google in California"

result = ner(text)
print(result)

for entity in result:
    print(entity)
```

Task 3: Question Answering

```python
qa = pipeline("question-answering")

context = "Transformers are models that process all words in parallel."
question = "What do transformers process?"

result = qa(question=question, context=context)

print(result)
```
