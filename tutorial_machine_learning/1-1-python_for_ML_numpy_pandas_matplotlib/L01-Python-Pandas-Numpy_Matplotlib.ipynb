```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# For dataset
from sklearn.datasets import load_iris
```

```python
# LOAD DATASET
iris = load_iris()

# Convert to pandas DataFrame
df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)

# Add target column
df["species"] = iris.target

print(df)
```

# PANDAS ESSENTIALS

```python
# First look at data

# print(df.head(10))
# print(df.sample(10))
print(df.tail(10))
```

```python
# Dataset info (very important for ML)

print(df.info())

# Info: count, data type, memory usage
```

```python
# Statistical summary

print(df.describe())
```

```python
# Check class distribution (classification relevance)

print(df["species"].value_counts())
```

```python
print(df.head(3))
```

```python
# Rename columns (cleaning)

df.columns = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "species"
]

print(df.head(5))
```

```python
# Filtering rows: all rows where petal_length > 4

mask = df["petal_length"] > 4
df_large_petal = df[mask]

print(df_large_petal.head())
```

```python
# Creating new feature (feature engineering)

df["petal_area"] = df["petal_length"] * df["petal_width"]
print(df.head(5))
```

```python
# remove a feature

df.drop(['petal_area'], axis=1, inplace=True)
print(df.head(6))
```

```python
# Check for missing values (example)

print(df.isnull().sum())
```

# NUMPY ESSENTIALS

```python
print(df.sample(6))
```

```python
# Convert DataFrame to NumPy array (common in ML)

X = df.drop("species", axis=1).values # features: all df except species
y = df["species"].values              # target

# Shape of data
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# print 1st 10 values of X and y

print(X[:10])
print(y[:10])
```

```python
# Basic NumPy operations
print("Mean of each feature:", np.mean(X, axis=0))
print("Standard deviation  :", np.std(X, axis=0))

# Min and Max (used in scaling)
print("Min values:", np.min(X, axis=0))
print("Max values:", np.max(X, axis=0))
```

```python
# Boolean masking

# Example: samples where sepal length > 6
mask = X[:, 0] > 6  # column idx=0 is sepal length

print("Shape of samples with sepal length > 6:", X[mask].shape)
print("First 5 samples with sepal length > 6:\n", X[mask][:5])
```

# MATPLOTLIB ESSENTIALS

```python
# Scatter plot (very common in ML exploration)

plt.figure()

plt.scatter(df["petal_length"], df["petal_width"])

plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.title("petal_length vs petal_width")
plt.show()
```

```python
# Scatter plot with class coloring

plt.figure()

for species in df["species"].unique():
    subset = df[df["species"] == species]
    plt.scatter(subset["petal_length"],
                subset["petal_width"],
                label=f"Class {species}"
               )

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width (Class-wise)")
plt.legend()
plt.show()
```

```python
# Histogram (feature distribution)

plt.figure()

plt.hist(df["sepal_length"], bins=20)

plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
# Boxplot for a single feature

plt.figure()
plt.boxplot(df["sepal_length"])

plt.ylabel("Sepal Length")
plt.title("Boxplot of Sepal Length")
plt.show()
```

```python
# Boxplots for multiple features (common ML EDA step)

plt.figure()
plt.boxplot([ df["sepal_length"], df["sepal_width"], df["petal_length"], df["petal_width"]],
     tick_labels=["Sepal Length",    "Sepal Width",     "Petal Length",    "Petal Width" ]
)

plt.ylabel("Value")
plt.title("Boxplot of Iris Features")
plt.show()
```
