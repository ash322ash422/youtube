```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
import pandas as pd

df = pd.read_csv("some_file.csv")
X = df[['test1', 'test2', 'test3']]
y = df['disease']

rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)

print(Counter(y_res))
```
