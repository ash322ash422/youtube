```python
import os
from newsapi import NewsApiClient
from textblob import TextBlob # need to install this on local machine
import json
```

```python
# 1. Initialize the NewsAPI client
# Note: Students should get their free API key from https://newsapi.org
NEWS_API_KEY = "YOUR_NEWSAPI"
newsapi = NewsApiClient(api_key=NEWS_API_KEY)

# 2. Fetch relevant financial news data
print("Fetching financial news for Apple Inc. (AAPL)...")

api_response = newsapi.get_everything(
    q='Apple stock OR AAPL',
    language='en',
    sort_by='relevancy',  # NOTE: other options include 'popularity' and 'publishedAt'
    page_size=5
)

articles = api_response.get('articles', [])
```

```python
print("\n==========================================")
print("     RAW JSON RESPONSE FROM NEWSAPI")
print("==========================================")
print(json.dumps(api_response, indent=4))
print("==========================================\n")
```

```python


# 3. Use the data in a Finance Domain context (Sentiment Analysis)
print("\n--- Market Sentiment Analysis Dashboard ---")

total_polarity = 0

for idx, article in enumerate(articles, 1):
    title = article['title']
    description = article['description'] or ""
    source = article['source']['name']

    # Combine title and description for a better text sample
    full_text = f"{title}. {description}"

    # Perform NLP Sentiment Analysis using TextBlob
    # Sentiment Polarity ranges from -1.0 (very negative) to +1.0 (very positive)
    analysis = TextBlob(full_text)
    sentiment_score = analysis.sentiment.polarity
    total_polarity += sentiment_score

    # Determine the visual label
    if sentiment_score > 0.05:
        sentiment_label = "🟢 BULLISH / POSITIVE"
    elif sentiment_score < -0.05:
        sentiment_label = "🔴 BEARISH / NEGATIVE"
    else:
        sentiment_label = "🟡 NEUTRAL"

    print(f"\n[{idx}] Source: {source}")
    print(f"Title: {title}")
    print(f"Sentiment Score: {sentiment_score:.2f} -> {sentiment_label}")

# 4. Calculate Aggregate Market Mood
if articles:
    average_sentiment = total_polarity / len(articles)
    print("\n==========================================")
    print(f"Aggregate Market Mood for AAPL: {average_sentiment:.2f}")
    if average_sentiment > 0.05:
        print("Final Trading Signal Suggestion: BUY / LONG (Overall Positive Sentiment)")
    elif average_sentiment < -0.05:
        print("Final Trading Signal Suggestion: SELL / SHORT (Overall Negative Sentiment)")
    else:
        print("Final Trading Signal Suggestion: HOLD (Overall Neutral Sentiment)")
    print("==========================================")
else:
    print("No articles found.")
```
