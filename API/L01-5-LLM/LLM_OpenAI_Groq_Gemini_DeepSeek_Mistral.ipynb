```python
import os
from openai import OpenAI

# Initialize the client.
# Idealy you should read the OPENAI_API_KEY from environment variable.
API_KEY = "key_here"
client = OpenAI(api_key=API_KEY)

# Create a chat completion request
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Hello! Tell me a short joke about coding."}
    ],
    temperature=0.7
)

# Print the model's response
print("AI Response:", response.choices[0].message.content)
```

# Gemini

```python
import os
from google import genai
from google.genai import types

# Initialize the client.
# Ideally, you should read the GEMINI_API_KEY from an environment variable.
API_KEY = "key_here"

client = genai.Client(api_key=API_KEY)

# Define system instruction and user prompt separately for Gemini
config = types.GenerateContentConfig(
    system_instruction="You are a helpful assistant.",
    temperature=0.7
)

# Create a generation request using the flagship model
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Hello! Tell me a short joke about coding.",
    config=config
)

# Print the model's response
print("AI Response:", response.text)
```

```python
import os
from mistralai.client import Mistral

# Initialize the official Mistral client
API_KEY = "key_here"
client = Mistral(api_key=API_KEY)

# Create a native chat completion request
response = client.chat.complete(
    model="mistral-large-latest",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Hello! Tell me a short joke about coding."}
    ],
    temperature=0.7
)

# Print the model's response using Mistral's native object structure
print("AI Response:", response.choices[0].message.content)
```
