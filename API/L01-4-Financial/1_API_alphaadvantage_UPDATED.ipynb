# 4) Technical Data
```
Relative Strength Index. - RSI


Scale: The RSI oscillates on a fixed scale between 0 and 100.

Overbought (>= 70): An RSI value at or above 70 suggests that an asset is overvalued and may be primed for a trend reversal or corrective price pullback.

Oversold (<= 30): An RSI value at or below 30 suggests that an asset is undervalued or oversold, indicating a potential buying opportunity or price bounce
```

## How Traders Use It
```
Reversal Signals: Selling when the RSI drops back below 70 from an overbought state, or buying when it climbs back above 30 from an oversold state

Divergence: If the asset price hits a new high, but the RSI fails to make a new high, it signals weakening momentum, suggesting the price might drop.

Trend Confirmation: An RSI consistently remaining above 50 generally indicates a strong bullish uptrend, while staying below 50 indicates a bearish downtrend [1].



````python
TICKER='AAPL'
url = (
    f"https://www.alphavantage.co/query?"
    f"function=RSI"
    f"&symbol={TICKER}"
    f"&interval=daily"
    f"&time_period=14"
    f"&series_type=close"
    f"&apikey={ALPHAADVANTAGE_API_KEY}"
)

data = requests.get(url).json()
````

````python
data
````

````python
# Lets see the keys

data.keys()
````

````python
data['Meta Data']
````

````python
data['Technical Analysis: RSI']
````

````python
# Step 1: Load dictionary using 'index' orientation
df = pd.DataFrame.from_dict(data['Technical Analysis: RSI'], orient='index')

# Step 2: Convert the date index to actual datetime objects
df.index = pd.to_datetime(df.index)
df.index.name = 'Date'

# Step 3: Convert the RSI column from text to numeric floats
df['RSI'] = pd.to_numeric(df['RSI'])

# Step 4: Sort chronologically (oldest date to newest date)
df = df.sort_index()

print(df)
````
