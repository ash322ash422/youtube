# API: Application Programming Interface
- It allows one software (Python) to talk to another software (Server)
- API endpoint: https://api.open-meteo.com/v1/forecast

```python
import requests

# Weather API endpoint (Open-Meteo)
url = "https://api.open-meteo.com/v1/forecast"

# Query parameters
params = {
    "latitude": 28.61,     # Delhi latitude
    "longitude": 77.20,    # Delhi longitude
    "current_weather": True
}


# Send GET request
response = requests.get(url, params=params)

if response.status_code == 200:
    print("Weather API Status Code:", response.status_code)
    
    # Convert response to JSON
    data = response.json()
    
    print("\nCurrent Weather Data:")
    print(data["current_weather"])

else:
    print("Error occurred")
    print("Status Code:", response.status_code)
```
