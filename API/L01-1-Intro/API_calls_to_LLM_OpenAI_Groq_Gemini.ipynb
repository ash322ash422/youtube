## Temperature: 

The "Creativity" SliderTemperature controls the flatness of the word probabilities.

- Low Temperature (e.g., 0.2): Makes the model highly predictable and focused. It will almost always choose the most likely next word. Perfect for coding, factual answers, and data extraction.

- High Temperature (e.g., 1.0 or higher): Flattens the probabilities, making less likely words more common. This introduces risks of "hallucinations" or typos, but increases creative variety. Perfect for brainstorming, poetry, and fiction.

## Top-p: The "Pool of Choices" Filter 

Also known as Nucleus Sampling, top_p controls how many words the AI considers by looking at their cumulative probability.


### How it works: 
- If you set top_p=0.15, the AI ranks all possible next words from most likely to least likely. It then cuts off the list as soon as the top words' combined probability reaches 15%. It only chooses from that tiny group.
- Low Top-p (e.g., 0.1): Only considers the absolute top fraction of words. It eliminates the "long tail" of unusual or unexpected words entirely.
- High Top-p (e.g., 0.95 or 1.0): Keeps almost all words in play, allowing the model to choose unusual or surprising vocabulary if the temperature allows it
