```python
import numpy as np
```

```python
# example 1 of clip: 
# Lets say we know that a sensor should return values between 20 - 150

# The sensor returned following values. Some of them are not in range 20 - 150
readings_sensor = np.array([
    14, 15,                  # this 2 would get clipped to 20 
    25, 30, 40, 45, 65, 85,  # these values remain same
    155, 160, 175            # these 3 would get clipped to 150
])
print(f"sensor readings: {readings_sensor}")

# Step 3: Clip the values between 20 and 150
safe_readings  = np.clip(readings_sensor, 20, 150)
print(f"safe_readings  : {safe_readings }") 

# NOTE:Values 10 and 15 would get clipped to 20. Values 155, 160, 175, 200 would clipped to 150 
```

```python
# Example 2 of clip: I know from experience that my values should be between 40 and 140

# Step 1: Generate 250 normally distributed data with mean=85 and StdDev=25
np.random.seed(42)  # for reproducibility
random_normal = np.random.normal(loc=85, scale=25, size=250)
random_normal = np.sort(random_normal).astype(int)
print(f"Numbers (Before clip): {random_normal}")

# Step 2: Clip the data between 40 and 140, then convert to int
clipped_data = np.clip(random_normal, 40, 140).astype(int)
clipped_data = np.sort(clipped_data)
print(f"Numbers (After clip): {clipped_data}")
```

```python
######## Step 3:  Lets visualize before and after clipping ###############
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Original data with KDE
plt.subplot(1, 2, 1)
sns.histplot(random_normal, bins=30, kde=True, color='skyblue')
plt.title('Original Data (Normal Dist.)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Clipped data with KDE
plt.subplot(1, 2, 2)
sns.histplot(clipped_data, bins=30, kde=True, color='salmon')
plt.title('Clipped & Converted Data (40 to 140)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
