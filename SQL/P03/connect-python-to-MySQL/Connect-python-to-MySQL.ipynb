```python
import mysql.connector
```

```python
# Configuration details for your MySQL database
config = {
    'user': 'root',  # Replace with your MySQL username
    'password': 'mysql@123',  # Replace with your MySQL password
    'host': 'localhost',      # or 127.0.0.1, 132.12.34.27
    'database': 'playground_db' # Replace with your database name
}


try:
    cnx = mysql.connector.connect(**config)
    print("Connection established successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
```

```python
# 1) Create a table employees
from mysql.connector import Error
import datetime

try:
    cnx = mysql.connector.connect(**config)
    if cnx.is_connected():
        # Create a cursor object
        cursor = cnx.cursor()

        # Define the SQL statement for creating the employees table
        create_table_query = """
        CREATE TABLE IF NOT EXISTS employees (
            id INT AUTO_INCREMENT PRIMARY KEY,
            fname VARCHAR(50) NOT NULL,
            lname VARCHAR(50) NOT NULL,
            dob DATE
        )
        """
        cursor.execute(create_table_query)
        print("Table 'employees' ensured to exist (created if not present).")

except Error as e:
    print(f"Error: {e}")

finally:
    # Close the cursor and connection
    if cnx and cnx.is_connected():
        cursor.close()
        cnx.close()
        print("\nMySQL connection is closed.")
```

## Verify that table has been created

```python
# 2) Populate the table with sample values

try:
    cnx = mysql.connector.connect(**config)
    if cnx.is_connected():
        # Create a cursor object
        cursor = cnx.cursor()

        # SQL statement for inserting data
        insert_query = "INSERT INTO employees (fname, lname, dob) VALUES (%s, %s, %s)"

        # Sample data as a list of tuples
        employees_to_add = [
            ('John', 'Doe', datetime.date(1990, 5, 15)),
            ('Jane', 'Smith', datetime.date(1985, 10, 20)),
            ('Peter', 'Jones', datetime.date(1992, 3, 10))
        ]

        # Use executemany() to insert multiple rows efficiently
        cursor.executemany(insert_query, employees_to_add)

        # Commit the changes (essential for data modification operations)
        cnx.commit()
        print(f"{cursor.rowcount} records inserted successfully.")

        
except Error as e:
    print(f"Error: {e}")

finally:
    # Close the cursor and connection
    if cnx and cnx.is_connected():
        cursor.close()
        cnx.close()
        print("\nMySQL connection is closed.")
```

```python
# 3. Verify the data insertion by selecting all rows

try:
    cnx = mysql.connector.connect(**config)
    if cnx.is_connected():
        # Create a cursor object
        cursor = cnx.cursor()

        select_query = "SELECT * FROM employees"
        cursor.execute(select_query)
        records = cursor.fetchall()

        print("\nData in the 'employees' table:")
        for row in records:
            print(row)
            
except Error as e:
    print(f"Error: {e}")

finally:
    # Close the cursor and connection
    if cnx and cnx.is_connected():
        cursor.close()
        cnx.close()
        print("\nMySQL connection is closed.")
```

# Read from a CSV file.

```python
import csv

try:
    cnx = mysql.connector.connect(**config)

    if cnx.is_connected():
        cursor = cnx.cursor()

        insert_query = "INSERT INTO employees (fname, lname, dob) VALUES (%s, %s, %s)"

        employees_to_add = []

        # Read data from CSV
        with open("data.csv", "r") as file:
            csv_reader = csv.DictReader(file)

            for row in csv_reader:
                fname = row["fname"]
                lname = row["lname"]

                # Convert string date to datetime.date object
                dob = datetime.datetime.strptime(row["dob"], "%Y-%m-%d").date()

                employees_to_add.append((fname, lname, dob))

        # Insert records into MySQL
        cursor.executemany(insert_query, employees_to_add)

        cnx.commit()
        print(f"{cursor.rowcount} records inserted successfully.")

except Error as e:
    print(f"Error: {e}")

finally:
    if cnx and cnx.is_connected():
        cursor.close()
        cnx.close()
        print("\nMySQL connection is closed.")
```

```python
# Verify the data insertion by selecting all rows

try:
    cnx = mysql.connector.connect(**config)
    if cnx.is_connected():
        # Create a cursor object
        cursor = cnx.cursor()

        select_query = "SELECT * FROM employees"
        cursor.execute(select_query)
        records = cursor.fetchall()

        print("\nData in the 'employees' table:")
        for row in records:
            print(row)
            
except Error as e:
    print(f"Error: {e}")

finally:
    # Close the cursor and connection
    if cnx and cnx.is_connected():
        cursor.close()
        cnx.close()
        print("\nMySQL connection is closed.")
```

```python
import pandas as pd

try:
    cnx = mysql.connector.connect(**config)

    if cnx.is_connected():
        cursor = cnx.cursor()

        insert_query = "INSERT INTO employees (fname, lname, dob) VALUES (%s, %s, %s)"

        
        df = pd.read_csv("data.csv")
        df['dob'] = pd.to_datetime(df['dob']).dt.date
        employees_to_add = list(df.itertuples(index=False, name=None))
        
        # Insert records into MySQL
        cursor.executemany(insert_query, employees_to_add)

        cnx.commit()
        print(f"{cursor.rowcount} records inserted successfully.")

except Error as e:
    print(f"Error: {e}")

finally:
    if cnx and cnx.is_connected():
        cursor.close()
        cnx.close()
        print("\nMySQL connection is closed.")
```
