# Test API

```python
import requests
import json

# Replace with your actual API Gateway URL
url = "https://19sbcltjdc.execute-api.ap-south-1.amazonaws.com/analyze"

headers = { "Content-Type": "application/json" }
payload = { "text": "I love learning API. I like AWS Lambda because it is easy."}
# payload = { "text": "I hate learning API. AWS Lambda makes me angry."}

# Make POST request
response = requests.post(url, headers=headers, data=json.dumps(payload))

# Print response
print("Status Code:", response.status_code)
print("Response JSON:", response.json())
```
