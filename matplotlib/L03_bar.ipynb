## Bar
A **bar plot** is used to **compare quantities across categories**.

---

### What a Bar Plot Is Best For

**1. Comparing discrete categories**

When your x-axis has **names / labels**, not continuous numbers.

Examples:

* Sales by **product**
* Number of students in each **class**
* Population of different **cities**
* Count of orders per **month**

---

**2. Showing differences clearly**

Bar plots make it very easy to see:

* Which category is **highest**
* Which is **lowest**
* How big the **gap** is between categories

---

### When You Should Use a Bar Plot

Use a bar plot when:

* Data is **categorical**
* You want **comparison**
* Order matters less than **magnitude**

---

```python
import matplotlib.pyplot as plt
```

```python
# 1) 
x = ["class9", "class10", "class11", "class12"] # class
y = [ 30,         40,           35,        27]  # number of children.

plt.bar(x, y, color="green")

plt.show()
```

```python
# 2) Different colors on bars
x      = ["class9", "class10", "class11", "class12"] # class
y      = [ 30,         40,           35,        27]  # number of children,
colors = ["red",    "blue",      "green",  "orange"]

plt.bar(x, y, color=colors)

plt.show()
```

```python
# 3) Draw horizonatal bars when you have too many classes/categories.

x      = ["class5", "class6", "class7", "class8", "class9", "class10", "class11", "class12"] # class
y      = [ 30,         40,       35,        27,       30,         40,        35,      27]

plt.barh(x, y, color=colors) # changed bar -> barh
plt.show()
```

```python
# 4) Now lets change the width
x = ["class9", "class10", "class11", "class12"] # class
y = [30,         40,           35,        27]  # number of children

plt.bar(x, y, color="green", width = 0.1)
plt.show()
```

```python
# Advanced application of bar: Bar plot for comparision
# 1) 
import numpy as np

x = np.array(["class9", "class10", "class11", "class12"]) # 4 classes in  chools

# male/female counts for two classes
male_counts   = np.array([3, 4, 3, 7])
female_counts = np.array([2, 3, 5, 8])

# Bar width and positions
bar_width = 0.30
x_pos = np.arange(len(x))

print(x_pos)
print(x_pos - bar_width/2)
print(x_pos + bar_width/2)
```

```python
# 2) 

# Plotting grouped bars
plt.bar(x_pos - bar_width/2, male_counts,   width=bar_width, label='Male',   color='blue')
plt.bar(x_pos + bar_width/2, female_counts, width=bar_width, label='Female', color='orange')

# X-axis labels and legend
plt.xticks(x_pos, x)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Count of People by Class")
plt.legend()
plt.tight_layout()

plt.show()
```
