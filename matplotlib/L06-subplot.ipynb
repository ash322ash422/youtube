## subplot()

A **subplot** is used to **display multiple plots in a single figure** so you can **compare them side-by-side or in a grid**.

---

### What `subplot()` Is Best For


**Saving space and telling one story**
Instead of opening many figures, subplots:

* Keep related plots together
* Make reports and dashboards cleaner


```python
# import library

import matplotlib.pyplot as plt 
```

```python
# I want 6 plots in tabular form with 2 rows and 3 columns 

# STEP1) Draw 6 plots in tabular form with 2 rows and 3 columns 

plt.subplot(2, 3, 1) # 2 rows, 3 cols, 1st subplot
# plot code here

plt.subplot(2, 3, 2) 
# plot code here

plt.subplot(2, 3, 3) 
# plot code here

plt.subplot(2, 3, 4) 
# plot code here

plt.subplot(2, 3, 5) 
# plot code here

plt.subplot(2, 3, 6) 
# plot code here

plt.show()
```

```python
# 2) Now lets fill in with some visuals
import numpy as np

plt.subplot(2, 3, 1) # 2 rows, 3 cols, 1st subplot
x = np.linspace(-4, +4, 30)
y = 5*(x**2) -3*x + 1
plt.plot(x, y)


plt.subplot(2, 3, 2) 
x = np.linspace(-4, +4, 30)
y = 5*(x**3) -5*x + 1
plt.plot(x, y)

plt.subplot(2, 3, 3) 
x = np.linspace(-4, +4, 30)
y = (x**3) -x + 5
plt.plot(x, y)

plt.subplot(2, 3, 4) 
x = np.linspace(-4, +4, 30)
y = (x**3) +x + 5
plt.plot(x, y)

plt.subplot(2, 3, 5) 
x = np.linspace(-4, +4, 30)
y = 5*(x**3) -5*x + 1
plt.plot(x, y)

plt.subplot(2, 3, 6) 
x = np.linspace(-4, +4, 30)
y = -5*(x**2) -5*x +1
plt.plot(x, y)

plt.show()
```
