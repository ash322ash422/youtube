# Histogram

A **histogram** is used to **understand the distribution of numerical data**.

*“How are values spread?”*
*“Where do most values lie?”*

---

### What a Histogram Is Best For

**1. Understanding data distribution**
It shows:

* Where data is **concentrated**
* Where data is **sparse**
* Overall **shape** of the data

Examples:

* Distribution of exam marks
* Distribution of salaries, ages, sensor readings

---

**2. Identifying patterns**
Histograms help you see:

* **Normal distribution** (bell-shaped)
* **Skewed data** (left or right skew)
* **Outliers**
* **Multiple peaks** (bimodal data)

---

**3. Frequency of value ranges (bins)**
Instead of exact values, histograms group data into **intervals (bins)**.

Example:

* Marks between 60–70
* Marks between 70–80
* Marks between 80–90

---

### When You Should Use a Histogram

Use a histogram when:

* Data is **continuous numerical**
* You want to see **spread and shape**
* You care about **frequency**, not individual points

---


```python
import matplotlib.pyplot as plt
```

```python
# 1) Generate data
import numpy as np

# Following could be age of employees in a company
data = np.array([51, 55, 55, 55, 56, 56, 57, 57, 57, 57, 58, 58, 59, 59, 59,
    59, 59, 60, 60, 60, 61, 61, 61, 61, 61, 61, 62, 62, 62, 62, 62, 62, 62,
    62, 62, 62, 63, 63, 63, 63, 63, 63, 63, 63, 63, 63, 63, 63, 64, 64, 64,
    64, 64, 64, 65, 65, 65, 65, 65, 65, 65, 65, 66, 66, 66, 66, 66, 66, 66,
    66, 66, 66, 66, 66, 66, 67, 67, 67, 68, 68, 68, 68, 69, 69, 69, 69, 69,
    69, 69, 70, 70, 70, 71, 72, 72, 72, 72, 72, 72, 74]
)

data = np.sort(data.astype(int))
print("data:\n", ", ".join(map(str, data)))
```

```python
# 2) Plot the histogram with only 2 bins

bin_count=2
counts, bins, patches = plt.hist(data, bins=bin_count)

plt.show()

print(f"bins  : {np.round(bins,1)}")
print(f"counts: {counts}")
```

```python
# 3) Prettify

bin_count = 2
counts, bins, patches = plt.hist( data, bins=bin_count,
    edgecolor='green',         # green edges for contrast
    color='gold'               # brighter color for visibility
)

plt.title('Histogram of Ages', color='green')
plt.xlabel('age (years)', color='green')
plt.ylabel('Frequency', color='green')
plt.tick_params(colors='green')

plt.show()

print(f"bins  : {np.round(bins,1)}")
print(f"counts: {counts}")
```

## observation about bin size
Here **bin_width = (max_value - min_value)/num_of_bins = (74-51)/2 = 11.5**.
- The bin starts from minimum value, which is 51, and then increase by 11.5 units. 
- The bins go from 51 -> 62.5 -> 74
- bin 51-62.5 contains 36 elements, bin 62.5-74 contains 64 elements

**Lets use above and increase the number of bins**

```python
# 4) Lets increase the bin count to 4

bin_count = 4
counts, bins, patches = plt.hist( data, bins=bin_count,
    edgecolor='green',         # green edges for contrast
    color='gold'               # brighter color for visibility
)

plt.title('Histogram of Ages', color='green')
plt.xlabel('age (years)', color='green')
plt.ylabel('Frequency', color='green')
plt.tick_params(colors='green')

plt.show()

print(f"bins  : {np.round(bins,1)}")
print(f"counts: {counts}")
```

```python
# 5)  Lets increase the bin count to 10

bin_count = 10
counts, bins, patches = plt.hist( data, bins=bin_count,
    edgecolor='green',         # green edges for contrast
    color='gold'               # brighter color for visibility
)

plt.title('Histogram of Ages', color='green')
plt.xlabel('age (years)', color='green')
plt.ylabel('Frequency', color='green')
plt.tick_params(colors='green')

plt.show()

print(f"bins  : {np.round(bins,1)}")
print(f"counts: {counts}")
```

```python
# Now increase the data to 10000 and bins to 100
import numpy as np

np.random.seed(42)

# mean = 65, std. dev. = 5, size =10000
data = np.random.normal(65, 5, 10000)  # heights in inches
bin_count = 100

counts, bins, patches = plt.hist( data, bins=bin_count,
    edgecolor='green',         # green edges for contrast
    color='gold'               # brighter color for visibility
)

plt.title('Histogram of Ages', color='green')
plt.xlabel('age (years)', color='green')
plt.ylabel('Frequency', color='green')
plt.tick_params(colors='green')

plt.show()
```
