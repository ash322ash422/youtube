```python
# import library

import matplotlib.pyplot as plt # create alias plt
```

```python
x = [1,  2,  3,  4,  5,  6 ] # day
y = [10, 15, 10, 17, 20, 25] # temp
```

```python
plt.plot(x, y)
plt.show() # render
```

```python
# add marker with colors
plt.plot(x, y, 'g*') # try g+ g* bo
plt.show()
```

```python
# add line style
plt.plot(x, y, 'b*--')
plt.show()
```

```python
plt.plot(x, y, 'go--', linewidth=2, markersize=12)
plt.show()
```
