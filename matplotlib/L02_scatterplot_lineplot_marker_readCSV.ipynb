```python
# 1) Now I want to plot graph of 2 different pairs
# Study hours vs Test scores for Alice and Bob

# Alice's study hours and scores 
x_alice = [2,  3,   4,  5,  6,  7,  8]
y_alice = [50, 55, 58, 60, 65, 67, 68]

# Bob's study hours and scores 
x_bob = [2,   3,  4,  5,  6,  7,  8]
y_bob = [55, 60, 65, 70, 75, 78, 82]

plt.figure(figsize=(5, 5)) # size = (length, width)

# Plotting data
plt.scatter(x_alice, y_alice, color='green',   marker='o')
plt.scatter(x_bob,   y_bob,   color='magenta', marker='s')

plt.show()
```

```python
# 2) Prettify: Now I am going to add labels and display it.
# Study hours vs Test scores for Alice and Bob

# Alice's study hours and scores 
x_alice = [2,  3,   4,  5,  6,  7,  8]
y_alice = [50, 55, 58, 60, 65, 67, 68]

# Bob's study hours and scores 
x_bob = [2,   3,  4,  5,  6,  7,  8]
y_bob = [55, 60, 65, 70, 75, 78, 82]

plt.figure(figsize=(5, 5)) # size = (length, width)

# Plotting data
plt.scatter(x_alice, y_alice, color='green',   marker='o', label='Alice')
plt.scatter(x_bob,   y_bob,   color='magenta', marker='s', label='Bob')

plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")
plt.title("Relationship Between Study Time and Test Performance")
plt.legend() # This is responsible for displaying labels
plt.grid(True)
plt.ylim(0, 100) # This is responsible for limiting y-axis
plt.xlim(1, 9)   # This is responsible for limiting x-axis

plt.show()
```

# Line plot

```python
# 1) line plots

# step1: plot (20,40), (25,55), (50,60)
x = [20, 25, 50] # age
y = [40, 55, 60] # wt

plt.plot(x, y)

plt.show() # render
```

```python
# 3) Different styles of line plot
x = [20, 25, 50] # age
y = [40, 55, 60] # wt

# You can also use the shortcut string notation parameter to specify the marker.
# This parameter is also called fmt, and is written with this syntax: marker|line|color
plt.plot(x, y, 'D-.g') # NOTE other examples: o--b o--r o:g o:r o-r o-.r 'o--g'

plt.show() # render
```

```python
# 4) marker size is 20

x = [20, 25, 50] # age
y = [40, 55, 60] # wt

plt.plot(x, y, marker = 'o', ms = 20) # ms is marker size

plt.show() # render
```

```python
# (SKIP) create a pair scatter matrix

# Sample data
data = np.array([
    [25, 50000, 700],
    [30, 60000, 720],
    [45, 80000, 680],
    [35, 75000, 690],
    [23, 52000, 710]
])

labels = ['Age', 'Income', 'Credit Score']
num_vars = data.shape[1]

fig, axes = plt.subplots(nrows=num_vars, ncols=num_vars, figsize=(8, 8))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

for i in range(num_vars):
    for j in range(num_vars):
        ax = axes[i, j]
        if i == j:
            ax.hist(data[:, i], bins=10, color='lightgray', edgecolor='black')
        else:
            ax.scatter(data[:, j], data[:, i], alpha=0.7)
        if i == num_vars - 1:
            ax.set_xlabel(labels[j], fontsize=9, rotation=45)
        else:
            ax.set_xticks([])
        if j == 0:
            ax.set_ylabel(labels[i], fontsize=9)
        else:
            ax.set_yticks([])

plt.suptitle("Manual Scatter Matrix", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# (SKIP) Lets plot different syle of markers
import numpy as np
import matplotlib.pyplot as plt

markers = [

'.', # point
',', # pixel
'o', # circle
'v', # triangle down
'^', # triangle up
'<', # triangle_left
'>', # triangle_right
'1', # tri_down
'2', # tri_up
'3', # tri_left
'4', # tri_right
'8', # octagon
's', # square
'p', # pentagon
'*', # star
'h', # hexagon1
'H', # hexagon2
'+', # plus
'x', # x
'D', # diamond
'd', # thin_diamond
'|', # vline

]

plt.figure(figsize=(13, 10))
samples = range(len(markers))


for i in samples:
    plt.plot([i-1, i, i+1], [i, i, i], label=markers[i], marker=markers[i], markersize=10)


# Annotation

plt.title('Matplotlib Marker styles', fontsize=20)
plt.ylim([-1, len(markers)+1])
plt.legend(loc='lower right')


plt.show()
```

```python
# SKIP : Lets look at some line styles

linestyles = ['-.', '--', 'None', '-', ':']

plt.figure(figsize=(8, 5))
samples = range(len(linestyles))


for i in samples:
    plt.plot([i-1, i, i+1], [i, i, i],
             label='"%s"' %linestyles[i],
             linestyle=linestyles[i],
             lw=4
             )

# Annotation

plt.title('Matplotlib line styles', fontsize=20)
plt.ylim([-1, len(linestyles)+1])
plt.legend(loc='lower right')


plt.show()
```

```python
# Annotate each points
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, y, color='blue')

# Annotate points with text
ax.text(x[0], y[0], "Point 1", ha="center", va="center", color="black")
ax.text(x[1], y[1], "Point 2", ha="center", va="center", color="black")
ax.text(x[2], y[2], "Point 3", ha="center", va="center", color="black")

# Set labels and title
ax.set_title('Scatter Plot with Annotations')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
# Lets add some text 

from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-1.5, 1.5, 30) # generate 30 points between -1.5 to +1.5
y = x**2

x_point = 0.8 # an x-point
y_point = x_point**2 # corresponding y point

#following plots a curve and a point
plt.plot(x, y, "g-", # plots a curve
         x_point, y_point, "ro" # plots a point
)

plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='red', horizontalalignment="center")
plt.text(x_point - 0.08, y_point, "Beautiful point", ha="right", weight="heavy")
plt.text(x_point, y_point, "x = %0.2f\ny = %0.2f"%(x_point, y_point), rotation=50, color='gray')

plt.show()
```
