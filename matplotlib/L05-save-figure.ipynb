# Save the figure

## Plot a curve and save the figure
$y = -1.5x^3 + 10$


```python
# Import matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1) Generate data points:

x = np.linspace(-4, 4, 30) # generate 30 points between -4 to +4
y = -1.5*(x**3) + 10 # y = -1.5(x^3) + 10

print("x:\n", x)
print("y:\n", y)
```

```python
# 2) Plot a curve and save the figure

plt.plot(x, y, marker="*")

# Save the figure BEFORE plt.show()
# Common formats: png, jpg, pdf, svg
plt.savefig("equation.png") # replaced plt.show()
```

## Saving with High Resolution (DPI)

```python
plt.figure(figsize=(6, 4))

plt.plot(x, y, marker="*")

plt.title("High Resolution Plot")

# dpi controls image quality (300 is publication quality)
plt.savefig("equation_high_res.png", dpi=300)

plt.show()
```

## Saving Multiple Formats

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker="*")
plt.title("Multiple Format Save")

plt.savefig("equation.png")
plt.savefig("equation.pdf")
plt.savefig("equation.svg")

plt.show()
```
