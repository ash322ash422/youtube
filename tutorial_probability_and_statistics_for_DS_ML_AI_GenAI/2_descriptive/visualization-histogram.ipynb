```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(42)
classA = np.random.normal(loc=75, scale=5, size=200)
classB = np.random.normal(loc=65, scale=10, size=200)

# Combine into a DataFrame for seaborn
import pandas as pd
df = pd.DataFrame({
    'Score': np.concatenate([classA, classB]),
    'Class': ['A'] * 200 + ['B'] * 200
})

# Plot overlapping histograms
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Score', hue='Class', kde=True, bins=20, stat='density', common_norm=False, alpha=0.6)
plt.title('Comparison of Exam Score Distributions Between Class A and Class B', fontsize=14)
plt.xlabel('Exam Score')
plt.ylabel('Density')
# plt.legend(title='Class')
plt.show()
```

## Observation about bin size
Here **bin_width = (max_value - min_value)/num_of_bins = (74-51)/2 = 11.5**.
- The bin starts from minimum value, which is 51, and then increase by 11.5 units. 
- The bins go from 51 -> 62.5 -> 74
- bin 51-62.5 contains 36 elements, bin 62.5-74 contains 64 elements

**Lets use above data and increase the number of bins**

```python
# step3: plot the histogram with 2, 4, 10 bins
bin_counts = [2, 4, 8]

for bin_count in bin_counts:
    counts, bin_edges = np.histogram(data, bins=bin_count)
    sns.histplot(data, bins=bin_edges, color='gold')
    plt.show()

    print("Bin edges:", bin_edges)
    print("Bin counts:", counts)
    print("Bin width:", bin_edges[1] - bin_edges[0])
    print("\n")
```

```python
# step4: Now increase the data size to 10000 and bins to 100
np.random.seed(42)

# mean = 65, std. dev. = 5, size =1000
data = np.random.normal(65, 5, 10000)
bin_count = 100

counts, bin_edges = np.histogram(data, bins=bin_count)
sns.histplot(data, bins=bin_edges, color='gold')
plt.show()
```

```python
# step5: Now lets add KDE curve to it
# KDE: Kernel Density Estimation-> It smooths data using a kernel at each point.

np.random.seed(42)

# mean = 65, std. dev. = 5, size =1000
data = np.random.normal(65, 5, 10000)  # heights in inches
bin_count = 100

counts, bin_edges = np.histogram(data, bins=bin_count)
sns.histplot(data, bins=bin_edges, color='gold', kde=True) 
plt.show()
```

## With histogram you can identify if the distribution is skewed

```python
# step1: Lets generate the data
data = np.round(np.random.exponential(scale=2, size=500), 2)
print("data=", data[:200])
```

```python
# step2: Plot and check if data is skewed
sns.histplot(data, kde=True)
plt.title('Skewed Distribution')
plt.show()
```

## With histogram you can identify if the distribution is uniform

```python
# step1: Get te data
data = np.round(np.random.uniform(2, 10, size=200), 2)
print("data=", data)
```

```python
# step2: Plot and check if data is uniform
sns.histplot(data, kde=True)
plt.title('Skewed Distribution')
plt.show()
```

## With histogram you can identify if the distribution is bimodal or unimodal

```python
# step1: The data
data = np.round(np.concatenate([np.random.normal(0,1,250), np.random.normal(5,1,250)]),2)
print("data=", data[:200])
```

```python
# step2: Now plot and check if its bimodal or unimodal
sns.histplot(data, kde=True)
plt.title('Bimodal Distribution')
plt.show()
```

# Histogram allows you to compare variance of 2 data

```python
# Create two normal distributions with same mean but different variances
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=2, size=1000)   
data2 = np.random.normal(loc=50, scale=10, size=1000)

# print(data1[:100])
# print(data2[:100])
```

```python
# step2: plot the histogram for both and identify low and high variance data

plt.figure(figsize=(10,5))
sns.histplot(data1, kde=True, color='skyblue', stat='density', bins=30) # low var
sns.histplot(data2, kde=True, color='salmon',  stat='density', bins=30, alpha=0.6)  # high var

# Labels and legend
plt.title('Effect of Variance on Histogram Shape', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
