```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('dark_background')
```

```python
# Lets plot just 3 points
```

```python
# Step 1: Weak positive relation
x = [2, 4, 3]
y = [5, 8, 4]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color='gold', marker="*", s=100)

plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

plt.xlim(0, 6)
plt.ylim(0, 10)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# Positive linear relation

```python
# Step 1: Weak positive relation
hours_studied = [1,  2,  3,   4,  5, 6, 2.5, 3.5, 4.5, 5.5, 1.5, 6.5]
scores =        [35, 40, 48, 70, 79, 77, 50, 50,  60,  70,  50,  95]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=hours_studied, y=scores, color='royalblue', s=100)

plt.xlabel('Hours', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
# Step 2: Strong positive relation
hours_studied = [1,   2, 3,  4,  5,  6, 2.5, 3.5, 4.5, 5.5, 1.5, 6.5]
scores        = [35, 45, 55, 60, 70, 90, 50, 58,  68,  80,  40,  95]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=hours_studied, y=scores, color='royalblue', s=100)

plt.xlabel('Hours', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
# Step 3: A point that do not fit the pattern
hours_studied = [1,   2, 3,  4,  5,  6, 2.5, 3.5, 4.5, 5.5, 1.5, 6.5]
scores        = [35, 45, 55, 60, 70, 46, 50, 58,  68,  80,  40,  95]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=hours_studied, y=scores, color='royalblue', s=100)

plt.xlabel('Hours', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

# Scatter plot help in identifying natural grouping of points

```python

np.random.seed(42)  # for reproducibility

# Cluster 1
x1 = np.random.normal(2, 0.5, 20)
y1 = np.random.normal(3, 0.5, 20)

# Cluster 2
x2 = np.random.normal(6, 0.5, 20)
y2 = np.random.normal(7, 0.5, 20)

# Cluster 3
x3 = np.random.normal(10, 0.5, 20)
y3 = np.random.normal(3, 0.5, 20)

# Combine all points
x = np.round(np.concatenate([x1, x2, x3]),1)
y = np.round(np.concatenate([y1, y2, y3]),1)

print("x:\n",", ".join(map(str, x)))
print("y:\n",", ".join(map(str, y)))


plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, s=100, color='royalblue')

plt.title('Scatter Plot Showing Natural Grouping of Points', fontsize=14, fontweight='bold')
plt.xlabel('Feature X', fontsize=12)
plt.ylabel('Feature Y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
