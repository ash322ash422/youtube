```python
# Importing LlamaIndex and loading documents
# Simplest way to create a Vector Store Index
# https://gpt-index.readthedocs.io/en/latest/how_to/vector_store_index.html
# SimpleDirectoryReader to load/Read files from a directory and treat them as text documents - https://developers.llamaindex.ai/python/framework/module_guides/loading/simpledirectoryreader/

from llama_index.core import VectorStoreIndex, SimpleDirectoryReader

documents = SimpleDirectoryReader('D:\\GENAIBatch\\langchain_new\\pdf_files').load_data()
```

```python
# Just To See the Loaded
documents
```

```python
# Creating the Vector Store Index
# It stores text chunks (nodes) as numerical vector embeddings in a vector database, allowing for efficient semantic search to find the most relevant information for an LLM-based query
# By default, it uses the OpenAI embedding model to convert text into vector representations, which can be used for similarity search and retrieval in LLM applications.
# But you can specify other embedding models if needed, and it also supports various vector database backends for storing and querying the embeddings.
# By default, LlamaIndex uses OpenAI's text-embedding-ada-002 model for generating embeddings, but you can specify other models if needed. It also supports various vector database backends for storing and querying the embeddings.

index = VectorStoreIndex.from_documents(documents, show_progress=True)
```

```python
# Query engine is a generic interface that allows you to ask question over your data.
# query engine takes in a natural language query, and returns a rich response
# It uses the underlying index to retrieve relevant information and generate a response based on the query.
# By default, the VectorStoreIndex uses cosine similarity to measure the distance between the query vector and the stored document vectors.
query_engine = index.as_query_engine()
```

```python
# check query engine
print(query_engine)
```

```python
# Run a query on the data managed by the index

response = query_engine.query("What is ReLU activation function?")
```

```python
# Print the response
print(response)
```

```python
#  pretty-print utility for displaying responses from LlamaIndex in a clean, human‚ÄĎreadable format

from llama_index.core.response.pprint_utils import pprint_response

pprint_response(response)
```

```python
# Pretty print with source nodes - Shows the source nodes (document chunks) that contributed to the answer.

pprint_response(response, show_source=True)
```

```python
# Additional imports (not used in this simple example, but useful for advanced usage)
# These imports are for more advanced retrieval and query engine customization
# vector index retriever - allows for more control over how documents are retrieved from the vector store
# retriever query engine - allows for custom query engines that use retrievers
# similarity postprocessor - allows for post-processing of query results based on similarity scores

from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.indices.postprocessor import SimilarityPostprocessor
```

```python
# Example of creating a custom retriever and postprocessor (not used in the simple example above)
# The VectorIndexRetriever allows you to specify how many top similar documents to retrieve based on the query, and the SimilarityPostprocessor allows you to filter results based on a similarity cutoff score.
# postprocessor can be used to filter out retrieved documents that do not meet a certain similarity threshold, ensuring that only relevant information is considered in the final response.
retriever = VectorIndexRetriever(index=index, similarity_top_k=5)
postprocessor = SimilarityPostprocessor(similarity_cutoff=0.80)
```

```python
# Creating a custom query engine with the retriever and postprocessor
query_engine = RetrieverQueryEngine(retriever=retriever, node_postprocessors=[postprocessor])
```

```python
# Example Query
response = query_engine.query("What is ReLU activation function?")
```

```python
print(response)
```

```python
# Pretty print with source nodes - Shows the source nodes (document chunks) that contributed to the answer.
pprint_response(response, show_source=True)
```

# Creating and storing the data (vectors) locally

```python
# Persisting and Loading the Index
# To avoid recreating the index every time, we can persist it to disk and load it later
import os
# import os.path to check if storage directory exists
import os.path

# import necessary LlamaIndex components for storage
from llama_index.core import (
    VectorStoreIndex,             # already imported above
    SimpleDirectoryReader,        # already imported above
    StorageContext,               # for managing storage context
    load_index_from_storage,       # for loading index from storage
)

# check if storage already exists
PERSIST_DIR = "./storage"

if not os.path.exists(PERSIST_DIR):
    
    # load the documents and create the index
    documents = SimpleDirectoryReader("D:\\GENAIBatch\\langchain_new\\pdf_files").load_data()
    index = VectorStoreIndex.from_documents(documents)
    
    # store it for later
    index.storage_context.persist(persist_dir=PERSIST_DIR)
    
else:
    # load the existing index
    storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_context)

# either way we can now query the index
query_engine = index.as_query_engine()
response = query_engine.query("What are transformers?")
print(response)
```

```python
# pretty-print utility for displaying responses from LlamaIndex in a clean, human‚ÄĎreadable format
pprint_response(response, show_source=True)
```

```python
from llama_index.core import Settings
from llama_index.llms.openai import OpenAI
from llama_index.embeddings.openai import OpenAIEmbedding

Settings.llm = OpenAI(
    model="gpt-4o-mini",   # fast + cost-efficient
    temperature=0.5
)

Settings.embed_model = OpenAIEmbedding(
    model="text-embedding-3-small"
)
```

```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader("D:\\GENAIBatch\\langchain_new\\pdf_files").load_data()
```

```python
import chromadb
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import StorageContext

# Persistent DB
chroma_client = chromadb.PersistentClient(path="./chroma_db")

collection = chroma_client.get_or_create_collection("rag_collection")

vector_store = ChromaVectorStore(chroma_collection=collection)

storage_context = StorageContext.from_defaults(vector_store=vector_store)
```

```python
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context
)
```

```python
query_engine = index.as_query_engine(similarity_top_k=4)

response = query_engine.query("Explain ReLU activation function")

print("\nAnswer:\n", response)
```
