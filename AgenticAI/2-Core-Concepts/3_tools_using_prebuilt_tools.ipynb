```python
from langchain_community.tools import WikipediaQueryRun, DuckDuckGoSearchRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain_experimental.tools.python.tool import PythonREPLTool
```

```python

#  Initialize Tools
wiki_wrapper = WikipediaAPIWrapper(
    top_k_results=2,   #  only few result
    doc_content_chars_max=1000  #  limit content size
)
wiki_tool = WikipediaQueryRun(api_wrapper=wiki_wrapper)

search_tool = DuckDuckGoSearchRun()
python_tool = PythonREPLTool()
```

```python
# 1. Test Wikipedia Tool

print("=== Wikipedia ===")
result = wiki_tool.invoke("Geoffrey Hinton")
print(result)
```

```python
# 2. Test DuckDuckGo Search Tool

print("\n=== Web Search ===")
result = search_tool.invoke("latest AI trends 2026")
print(result)
```

```python
# 3. Test Python REPL Tool
print("\n=== Python Execution ===")
result = python_tool.invoke("print(sum([10, 20, 30, 40]))")
print(result)
```
