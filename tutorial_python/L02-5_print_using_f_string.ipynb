# print using f-string
- f-string means **formatted string**
- Covenient method to print variable
- Make code readable and reduce formatting errors

```python
name = "Ash"          # String variable

print("Name is", name)
print(f"Name is {name}")
```

```python
age = 30              # Integer variable

print("Age is", age)
print(f"Age is {age}")
```

```python
height = 6.2          # Float variable

print("Height is", height)
print(f"Height is {height}")
```

```python
is_trainer = True     # Boolean variable

print("Is he/she a trainer ?", is_trainer)
print(f"Is he/she a trainer ? {is_trainer}")
```

```python
name = "Ash"
age = 25
height = 4.1

print("My name is", name, "age is", age, "and height is", height) # lots of commas
print(f"My name is {name} age is {age} and height is {height}")
```

# Perform computation in f-string

```python
age = 20

print("Age after 5 years:", age + 5)
print(f"Age after 5 years: {age + 5}")
```
