# String (str) Methods

```python
# Lets look at the string methods
print(dir(str))
# print(help(str))

# Important ones are the end without underscore _:
# capitalize', 'casefold', 'center', 'count', 'encode', 'endswith', 
# 'expandtabs', 'find', 'format', 'format_map', 'index', 'isalnum', 
# etc....
```

```python
# Length of string

text = "  python is EASY " 
print(text)
print(len(text))
```

```python
# Lets explore some String in-built methods

text = "  python is CoOL language. it is NOT a python snake " 
print(text)
print(text.lower())  
print(text.upper())  
print(text.count('o'))
print(text.find("is")) # index of 'is' : 9

# this is called chaining method. It counts all o regardless of upper/lower case.
print(text.lower().count('o')) 
```

```python
# remove 
text = "  python is CoOL language. it is NOT a python snake " 

new_text = text.replace("py", "my")
print(new_text) 
```

```python
# clean text
text = "ML@2025!!"
cleaned = text.replace("@", "").replace("!", "") # chaining
print(cleaned)
```

```python
# startswith
text = "Machine learning is powerful"

print(text.startswith("Machine"))   # True
print(text.startswith("learning"))  # False
```

```python
# endswith
filename = "model.pkl"

print(filename.endswith(".pkl"))  # True
print(filename.endswith(".csv"))  # False
```

```python
# Checking substring: Use 'in' operator because it is optimized
sentence = """I need to use a string representation of that in my program. I also
            need to be able to search it for substrings and the like - in other words,
            I need to be able to treat it like a normal sized string. I'll be trying to 
            find a lot of substrings so speed is necessary.
            My initial logic was to simply copy and paste the string directly into 
            I closed the program eventually)."""


b = "try" in sentence 
print(b)  # True

b = "rain" in sentence
print(b)  # False
```

```python
text = "  python is CoOL language.  it is NOT a python snake " 

print(text)
print(text.lstrip()) # Removes empty spaces from left
print(text.rstrip()) # Removes empty spaces from right
print(text.strip()) #Removes all empty spaces from left and right
```

```python
# Title
text = "  python is CoOL language.  it is NOT a python snake " 
print(text)

print(text.title())
```

```python
# sometimes you have to split a data by delimiter/seperator

# STEP1: Split based on ,
data = "2025-08-23,ashwani,34,000,34.78"
lst = data.split(",")

print(lst) # ['2025-08-23', 'ashwani', '34', '000', '34.78']
```

```python
#STEP2: Split based on ;
data = "2025-08-23;ashwani;34,000;34.78"
lst = data.split(";")

print(lst) 
```

```python
# join
words = ['2025-08-23', 'ashwani', '34,000', '34.78']

sentence = " ".join(words)

print(sentence)
```
