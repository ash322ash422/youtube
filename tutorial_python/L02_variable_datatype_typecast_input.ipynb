## 1. What Are Variables?

A **variable** is a container used to store data in memory.

Think of a variable as a **label** attached to a value.

### Why do we need variables?
- To store data
- To reuse values
- To make programs readable and flexible


````python
# Creating variables
age = 25
name = "Gary Patel"
height = 5.9

print(age)
print(name)
print(height)
````

## 2. Basic Data Types in Python

Python has several built-in data types.  
We will focus on:

- **Numbers or numeric** → int, float, complex
- **Strings** → text data
- **Booleans** → True / False


#### Numeric Types
**int** → Whole numbers: x = 10, -4, 7, 89

**float** → Decimal numbers: y = 10.5, -54.789, 3.567

**complex** → Complex numbers: z = 3 + 4j , -1 + 2j

````python
# Integer
x = 10
print(x, type(x))
````

````python
# Float
y = 3.14
print(y, type(y))
````

### Strings

Strings are used to store text.
They are written inside quotes (' ' or " ").


````python
message = "Hello Python"
name    = 'Student'

print(message)
print(name)
print(type(message))
````

### Booleans

Booleans represent **True** or **False**.
They are commonly used in conditions.


````python
is_python_easy = True
is_raining = False

print(is_python_easy, type(is_python_easy))
print(is_raining, type(is_raining))
````

## 3. Type Casting

Type casting means converting one data type into another.

Common functions:
- `int()`
- `float()`
- `str()`
- `bool()`


````python
# step1: 
x = "10"

print(type(x))
# print(x / 10) # ERROR
````

````python
# step2: now convert str into int
x = "10"
x = int(x)       # or float(x)

print(type(x))

print(x / 10) 
````

````python
# Integer to Float
a = 5
b = float(a)

print(b, type(b))
````

````python
# convert float -> int
x = 10.5 # float

x = int(x) # covert it into int
print(x, type(x))
````

````python
# Number to String
score = 95
score_str = str(score)

print(score_str, type(score_str))
````

## 4. input() – Basic User Interaction


The `input()` function is used to take input from the user.

### IMPORTANT
- `input()` **always returns a string**


````python
# prompt user to enter name
name = input("Enter your name: ")

print("Hello,", name)
````

````python
# Using Type Casting with input()
age = int(input("Enter your age: "))

new_age = age + 1
print("You will be", new_age, "next year.")
````
