# Basic Data Types in Python

Python has several built-in data types.  
We will focus on:
- **Numbers or numeric** → int, float, complex
- **Strings** → text data
- **Booleans** → True / False

#### Numeric Types
**int** → Whole numbers: x = 10, -4, 7, 89

**float** → Decimal numbers: y = 10.5, -54.789, 3.567

**complex** → Complex numbers: z = 3 + 4j , -1 + 2j

```python
# Integers have no decimal part
a = 10
b = -5
c = 0

print(a, b, c)
print(type(b))
```

```python
# Floating Point Numbers (float) have decimal part

price = 99.99
temperature = -12.5

print(price)
print(type(temperature))
```

```python
# Complex numbers: Used primarily in Engineering domain
z = 2 + 3j

print(z)
print(type(z))
```

### Strings

Strings are used to store text.
They are written inside quotes (' ' or " ").

```python
message = "Hello Python"
name    = 'Student are excited'

print(message)
print(name)
print(type(message))
```

### Booleans

Booleans represent **True** or **False**.
They are commonly used in conditions.


```python
is_python_easy = True
is_raining = False

print(is_python_easy, type(is_python_easy))
print(is_raining, type(is_raining))
```
