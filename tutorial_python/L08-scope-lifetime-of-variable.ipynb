## What is Scope?

**Scope** determines *where* a variable can be accessed in your code.

```python
# 2
```

## Local Scope (Inside a Function)

```python
### Example: step1) 

def greet():
    message = "Hello from inside the function" # local var
    print(message)

greet()

# Uncommenting below line will cause an error. You CANNOT access variable message
# print(message)
```

### Explanation

* `message` is a **local variable**
* Exists **only inside** the function
* Destroyed when function execution ends

```python
### Example: step2) 

def greet():
    message = "Hello from inside the function" # local var
    print(message)

greet()

message = "Now I am defined outside the function"
print(message)
```

## Lifetime of a Local Variable

```python
# example

def counter():
    x = 10
    print("Inside function:", x)

counter()

# x no longer exists here
# print(x)
```

- Variable `x` is **created when function starts**
- Variable `x` is **destroyed when function ends**

## Global Scope (Outside All Functions)

```python
### Example

city = "Moscow"   # Global variable

def show_city():
    print("City:", city)

show_city()
print("Accessing globally:", city)
```

### Explanation

* `city` is a **global variable**
* Accessible **inside and outside** functions


```python
# 4
```

## Modifying Global Variables (Common Pitfall)

```python
count = 0 # global var

def increase():
    # count = count + 1  # Error
    global count
    count = count + 1

increase()
increase()
print("Final count:", count)
```

### Key Rule

* To **modify** a global variable inside a function â†’ use `global`
