### Order of Arithmetic Operators

Suppose you have to evalute 
```
(3 + 1 * 4) * 20 / 5 + 1 - 2 ** 3
```


#### Python Operator Precedence (Highest to Lowest) - PEMDAS
Here’s the order in which Python evaluates operators:

- () → Parentheses

- ** (Exponentiation)

- *, / (Multiplication and Division)

- +, - (Addition and Subtraction)

- Operators with the same precedence are evaluated left to right (except **, which is right to left).

````python
# e.g. step1) 
a = 2
b = 3
c = 4

result = a + b * c

print(result)  # Output: 14
````

````python
# step2) Now I want addition to happen before multiplication
a = 2
b = 3
c = 4

result = (a + b) * c

print(result)  # Output: 20
````

````python
# e.g.

result = 2 ** 3 * 4  # = (2**3) * 4 = 8*4 = 32

print(result)  # Output: 32
````

````python
# e.g.

result = 10 + 6 / 2 * 3 - 1

print(result)  # Output: 18.0
````

````python
# e.g.

result = (3 + 1 * 4) * 20 / 5 + 1 - 2 ** 3

print(result)
````
