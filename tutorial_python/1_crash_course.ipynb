# 2. Control Flow Structures

````python
# if-elif-else
age = 29 # 10 15 29

if age < 13:
    print("Child")
    print("Age is", age)
    
elif age < 18:
    print("Teenager")
    print("Age is", age)
    
else:
    print("Adult")
````

````python
# input function

name   = input("Please enter your name:")
sport  = input("What is your favorite sport ?:")

if sport == "football":
    print("Hello", name, ". I also like football")
    
else:
    print("Hello", name, ". I wish I knew more about", sport)
````

````python
# step1: Use for loop for repetitive tasks: Run this finite number of times.

for i in range(3):
  print("Hello world")
  print("Today I will learn python")

print("+++++I am DONE+++++")
````

````python
# step2: Now what is this i ? i is an iterator
  
for i in range(4): # i is iterating over 0, 1, 2, 3
    print("Iteration no.:", i)
    print("Multiply iterator by 3:", 3 * i)
    
print("YEA...Finally I am out of for loop")
````

````python
# While loop: keep running until some condition is met

i = 0
while i < 3:
    print("The counter is:", i)
    print("Multiply counter by 3:", 3 * i)
    i += 1                 # increment i: i = i + 1

print("I am out of while loop")
````

# 3. Data Structures: List, Tuple, Set, Dictionary

````python
# List: square bracket []

age = [ 21, 55, 31, 64, 74, 31 ]

print("list is ", age)
print("element:", age[4]) # indexing starts with 0
print("type:", type(age))
````

````python
# Tuple: round bracket ()

age = ( 21, 55, 31, 64, 74, 31 )

print("tuple is ", age)
print("element:", age[4]) # indexing starts with 0
print("type:", type(age))
````

````python
# Set: curly braces {}

age = { 21, 55, 31, 64, 74, 31 }  # duplicates removed

print("set is ", age)
# print("element:", age[4]) # Error: indexing not possible
print("type:", type(age))
````

````python
# Dictionary --> { key: value } pairs

employee = {
    "name":       "Alice", 
    "age":        25,
    "salary":     30000,
    "is_married": True
}
print("The employee data:", employee)
print("salary:", employee['salary'])
print("type:", type(employee))
````

````python
# list of dictionary: list of employees
emp1 = { "name": "Harry", "age": 25, "salary": 30000 }
emp2 = { "name": "Khan",  "age": 53, "salary": 40000 }
emp3 = { "name": "Rekha", "age": 29, "salary": 94000 }

employees = [ emp1, emp2, emp3 ]
print(employees)
print(employees[0])
print("type:", type(employees))
````

# 5 File operations: Read or Write to a file

````python
# step1: Create and write to a file 
text1 = "Hello. This is a great day about file read and write operations. "
text2 = "Show me how to write to a file"

file = open("notes.txt", "w")

file.write(text1)
file.write(text2)

file.close()
````

````python
# step2: Read from a file 

file = open("notes.txt", "r")

content = file.read()

file.close()

print("The content of file:", content)
````

# 6. Error Handling

````python
# step1: Erroneous code
a = 0 # 0 or 5

result = 10 / a

print(result)
print("Moving on with code")
````

````python
# step2: You catch ZeroDivisionError error/exception
a = 0

try:
    result = 10 / a
    print(result)
except ZeroDivisionError:
    print("You can't divide by zero!")
    print("But, I will give you a pass")

print("Moving on with code")
````
