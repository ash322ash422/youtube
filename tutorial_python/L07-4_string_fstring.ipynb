# Python f-Strings
An f-string is a string that allows you to embed **Python expressions directly inside { }**.

- f-strings make your code cleaner, faster, and easier to read

```python
# step1) Without f-string
name = "Roger"

print("Hello", name)
```

```python
# step2) With f-string
name = "Roger"

print(f"Hello {name}")
```

## Using Multiple Variables

```python
name = "Bob"
age = 25

# Old way
print("Name: " + name + ", Age: " + str(age))

# Using f-string (recommended)
print(f"Name: {name}, Age: {age}")
```

## Formatting Floating-Point Numbers

```python
accuracy = 0.92345678

# Limit to 2 decimal places
print(f"Accuracy: {accuracy:.2f}")
```

## Embedding Expressions

```python
actual = 100
predicted = 92

# Absolute error calculation inside f-string
print(f"Absolute Error: {abs(actual - predicted)}")
```

```python
price = 100
qty = 92

print(f"Total price: {price * qty}")
```

```python
# Celsius -> Farenheit
c = 50 # Celsius

print(f"{c} Celsius is { (c * 9/5) + 32} Farenheit: ")
```

## Using f-strings with Lists (Data Summary)

```python
scores = [78, 85, 90, 88]

print(f"Number of samples: {len(scores)}")
print(f"Highest score: {max(scores)}")
print(f"Average score: {sum(scores) / len(scores):.2f}")
```
