# String (str): Concatenation

```python
# 1) You cannot add numeric types to string
a = "7" # str
b = 5   # int
# res = a + b # ERROR
```

```python
# 2) You can add string to string
a = "7"  # str
b = "5"  # str

res = a + b

print(res)
```

```python
# 3) You can add string to string: concatenation

a = "Hello, how are you. "    # str
b = "Thank you. I am fine"    # str
res = a + b 

print(res)
```

```python
# 4) You can add string to string: concatenation

a = "Hello, how are you. "    # str
b = "Thank you. I am fine"    # str
c = "\nI hope this concatenation  part is clear now"
res = a + b + c

print(res)
```

```python
# what about multiply ???
# step1) You can multiply numeric types to string: it gives string repetition
a = "873"

b = 5  # or 50 to create a big number 

res = a * b

print(res)
```

```python
# step2) Generate a big string that can be used as data for further use 
text = "Hello! It's a common English greeting used to start conversations, \
related to the French 'Hola'. "

giant_text = text * 5 # or 50

print(giant_text)
```
