### What is a Variable?

A variable is a name that refers to a **value stored in memory**.

### Why Do We Need Variables?
- Store data
- Reuse values

### Creating Variables in Python (Assignment)

```
variable_name = value
```

Python uses the equals sign (=) to assign a value to a variable.

````python
# Here I create an integer variable, text AKA string variable 
# and decimal AKA float variable 

x = 10
name = "Python"
rating = 4.8

print(x)
print(name)
print(rating)
````

````python
# 2
````

### Python is Dynamically Typed
- No need to declare data type
- Type is decided at runtime

````python
# Python does not require us to specify the data type.
# Python automatically understands the type based on the value.

x = 10
print(x)
print(type(x))
````

````python
# Python automatically understands the type based on the value.
# Change variable type here. Python allows that

x = 10.6
print(x)
print(type(x))
````

````python
# Python automatically understands the type based on the value.
# Change variable type here. Python allows that

x = "hello"
print(x)
print(type(x))
````

````python
# 3
````

````python
# We can perform calculation on variables. Below code is readable

price = 100
tax   = 18
total = price + tax

print(total)
````

````python
# 4
````

### Variable Naming Rules
#### Allowed:
- Letters
- Numbers
- Underscore (_)

#### Not Allowed:
- Starting with a number : 2_name not alowed
- Special characters: #, @, $, -, + etc are not allowed
- Python keywords: 

````python
# allowed
age = 25
student_name = "Rahul"
marks_1 = 90
_count = 3.9
````

````python
# not allowed 
# 1name = "Alex"     # Invalid
# total@price = 100  # Invalid
# total%price = 100  # Invalid
# class = 10         # Keyword
````

````python
# Python has reserved words called keywords, which cannot be used as variable names.
import keyword

print(keyword.kwlist)

true = 4
````

### Python is case-sensitive

````python
# age and Age are 2 different variables
age = 20
Age = 30

print(age)
print(Age)
# print(age + Age)
````

````python
# 6
````

### Multiple Variable Assignment

````python
# Assign multiple values in one line

a, b, c = 1, 2, 3

print(a, b, c)
print(a + b + c)
````

````python
# Same value to multiple variables:

x = y = z = 5

print(x, y, z)
````

````python
# 7
````

### Reassigning Variables

````python
# Variables can change their values.

count = 10
count = count + 1
print(count)
````
