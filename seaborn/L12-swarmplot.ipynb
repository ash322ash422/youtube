# swarmplot()
- Helps to avoid overlapping data points
- Ideal when you have small dataset: < 5000 data points 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
sns.swarmplot(
    data=tips,
    x="total_bill"
)

plt.title("Swarm Plot of Total Bill")
plt.show()
```

```python
# Compare swarm with KDE plot
# Create 2 rows, 1 column
fig, axes = plt.subplots(
    2, 1,
    figsize=(6, 5),
    sharex=True,          # Important: same x-axis
    gridspec_kw={'height_ratios': [1, 1]}
)

# Swarmplot (Top)
sns.swarmplot(
    data=tips,
    x="total_bill",
    ax=axes[0]
)
axes[0].set_title("Swarm Plot (Raw Data)")
axes[0].set_ylabel("")

# KDE (Bottom)
sns.kdeplot(
    data=tips,
    x="total_bill",
    fill=True,
    ax=axes[1]
)
axes[1].set_title("KDE Distribution")
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

# Observation:
- Most of the data is in between 10 and 25

```python
# Seperate total bill by day
sns.swarmplot(
    data=tips,
    x="day", # categorical
    y="total_bill"
)

plt.title("Swarm Plot of Total Bill per Day")
plt.show()
```

# Observation:
- For Sat, most of total bill was in range of 10 - 30


```python
# 1) tip by time
sns.swarmplot(data=tips,
              x='time', # should be categorical
              y='tip'   # should be numerical
)

plt.show()
```

```python
# 2) Add sex category
sns.swarmplot(data=tips,
              x='time', # should be categorical
              y='tip',  # should be numerical
              hue='sex' # should be categorical
)

plt.show()
```

# Observation:
- For lunch, it seems that more females give tips
- For dinner, more male give tips.
- Males give high tip in general

```python
# 3) Add sex category and seperate them. I do not want overlapping.

sns.swarmplot(data=tips,
              x='time', # should be categorical
              y='tip',  # should be numerical
              hue='sex', # should be categorical
              dodge=True # seperation
)
plt.show()
```

# Swarm + violin plot

```python
# 1)
sns.swarmplot(data=tips,
              y='tip',
              color='black'
)

sns.violinplot(data=tips,
               y='tip',
               color="lightgray",
               inner=None
)

plt.show()
```

```python
# tip by time
sns.swarmplot(data=tips,
              x='time',
              y='tip',
              color='black'
)

sns.violinplot(data=tips,
               x='time',
               y='tip',
               color="lightgray",
               inner=None
)

plt.show()
```

# swarmplot + boxplot

```python
# 1) total bill
sns.boxplot(
    data=tips,
    y="total_bill",
    color="lightgray"
)

sns.swarmplot(
    data=tips,
    y="total_bill",
    color="black"
)

plt.title("Boxplot + Swarm Plot")
plt.show()
```

```python
# 2)  total bill by day
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    color="lightgray"
)

sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill",
    color="black"
)

plt.title("Boxplot + Swarm Plot")
plt.show()
```
