# pairplot()
- Use pairplot when you want to explore relationships between multiple numerical features at once

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = sns.load_dataset("iris")

# Display first few rows
iris.head()
```

```python
# 
sns.pairplot( data=iris )

plt.title("Iris Dataset - Pairwise Relationships")

plt.show()
```

# observation:
- Strong positive linear relationship between petal length and petal width
- Weak linear relationship between sepal length and sepal width

```python
# Lets seperate based on species
sns.pairplot(
    data=iris,
    hue="species" # categorical
)

plt.show()
```

# Observation
- Setosa can easily be identified using petal length and petal width. For setosa, petal width = (0, 0.6) and petal length = (0, 2) 

```python
penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
sns.pairplot(  data=penguins )

plt.show()
```

# observation:
- Strong positive linear relationship between flipper length and body mass 

```python
# Show me the lower triangular data

sns.pairplot(penguins, corner=True)
plt.show()
```

```python
# Took 10 seconds

sns.pairplot(penguins, kind="kde")
plt.show()
```

```python
# lets seperate species

sns.pairplot(
    data=penguins,
    hue="species" # category
)

plt.show()
```

# Observation:
- Easy to seperate Adelie using bill depth and body mass
  Easy to seperate the 3 species using flipper length and bill length
