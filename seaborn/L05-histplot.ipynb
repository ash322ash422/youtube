# histplot()
- Understand how numerical data is distributed

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
tips['total_bill'].describe()
```

```python
# 1) lets look at distribution total_bill

sns.histplot(data=tips,  x='total_bill')

plt.show()
```

```python
# 2) lets look at distribution total_bill (on y-aixis)

sns.histplot(data=tips,  y='total_bill')

plt.show()
```

```python
# 3) lets look at distribution total_bill with KDE.

sns.histplot(data=tips,  x='total_bill',
    kde=True, # This plots the smooth curve
)

plt.show()
```

## observation:
* Most people in this dataset spent between $10$20 on their meal.
* A few people spent a lot more (up to ~$50), which the KDE curve will show as a long tail to the right.
* The distribution is **skewed to the right**

# binwidth and binrange

```python
#  My favorite binwidth and binrange

sns.histplot(data=tips,  x='total_bill',
             binwidth=2, binrange=(5, 50)
)

plt.show()
```

# Seperate by categories

```python
sns.histplot(data=tips, x="tip",
             hue="sex", # should be category
             kde=True
)

plt.show()
```

# observation:
- More men give tip than females. Thas because of more **area** under KDE for men
- The amount of tip given by male and females are almost same.

```python
iris = sns.load_dataset("iris")

# Display first few rows
iris.head()
```

```python
# 2) The distribution of petal length

sns.histplot(data=iris, x="petal_length",
             kde=True,
)

plt.show()
```

```python
# 2) The distribution of petal length by species

sns.histplot(data=iris, x="petal_length",
             hue="species",
             kde=True,
)

plt.show()
```

# Observation:
- Setosa has petal length between 1 and 2.5 . So we can easily identfy this species by looking at petal length
- Versicolor and virginica has little overlapping: Versicolor has petal length from 2.5 to 5.5 . Virginica has petal length from 4 to 7
- The amount of area under KDE is almost similar for 3 species. That means we have almost same number of setosa, versicolor, and virginica.
