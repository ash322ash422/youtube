# violinplot 
- Best suited for large datasets


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
# Compare numerical distr across categorical data

sns.violinplot(data=penguins,
               x="species",  # category
               y="flipper_length_mm" # numerical
)

plt.show()
```

# Observ:
- Adelie flipper length is mainly between 180 - 200

```python
sns.violinplot(data=penguins,
               x="species", # category
               y="flipper_length_mm", # numerical
               hue="sex",   # category
)

plt.show()
```

```python
sns.violinplot(data=penguins,
               x="species",
               y="flipper_length_mm",
               hue="sex",
               fill=False,
)

plt.show()
```

# Seperate Male / Female

```python
sns.violinplot(data=penguins,
               x="species",
               y="flipper_length_mm",
               hue="sex",
               split=True,
               inner="quart",
)

plt.show()
```

```python
sns.violinplot(data=penguins,
               x="species",
               y="flipper_length_mm",
               hue="sex",
               split=True,
               inner="quart",
               gap=0.1,
)

plt.show()
```
