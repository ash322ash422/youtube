# kdeplot()

Kernel Density Estimation (KDE) is a way to visualize the probability density of a continuous variable.

- It is a smooth curve and an alternative to a histogram
- Area under the curve = 1



### Avoid KDE when:
- Dataset is very small
- Data is discrete
- Distribution has extreme outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips['total_bill'].describe()
```

```python
# 1) 
sns.kdeplot(data=tips, x="total_bill")
plt.show()
```

```python
# 2) You can also fill
sns.kdeplot(data=tips, x="total_bill",
           fill=True,
           )
plt.show()
```

## observation:
* Most people in this dataset spent between $10$20 on their meal.
* A few people spent a lot more (up to ~$50), which the KDE curve will show as a long tail to the right.
* The distribution is **skewed to the right**

```python
iris = sns.load_dataset("iris")

# Display first few rows
iris.head()
```

```python
# simple density function
sns.kdeplot(data=iris, x='sepal_length')

plt.show()
```

# observation:
- The sepal length lies between 3.5 to 8.8.
- Most of sepal length is between 4.5 and 7

```python
# KDE plot for different species 

sns.kdeplot(data=iris, x='petal_length', fill=True,
           hue="species" # categorical
           )

plt.show()
```

# Observation:
- Setosa has petal length between 1 and 2.5 . So we can easily identfy this species by looking at petal length
- Versicolor and virginica has little overlapping: Versicolor has petal length from 2.5 to 5.5 . Virginica has petal length from 4 to 7
- The amount of area under KDE is almost similar for 3 species. That means we have almost same number of setosa, versicolor, and virginica.
