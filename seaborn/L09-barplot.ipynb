# barplot
- Used for categorical data + numerical
- Comparing categories: Tips on Fri, Sat, Sun 
- Showing counts or averages across various group 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
sns.barplot(
    data=tips,
    x="day", # category
    y="total_bill" # numerical
)

plt.title("Average Total Bill per Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

# Observation:
- Average tip is higher for Sat and Sun


Seaborn automatically calculates mean

Error bars show confidence interval

```python
sns.barplot(
    data=tips,
    x="day", # category
    y="total_bill", # numerical
    hue="sex" # categorical
)

plt.title("Average Total Bill per Day by Gender")
plt.show()
```

# Observation:
- On Fri, the average total bill for male was higher (20) than of that from female(14)
- total bill is always higher for males. The difference is not very high

```python
# show the sum of total bill

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    estimator=sum,
    errorbar=None
)

plt.title("Total Bill per Day (Sum)")
plt.show()
```

# Observ:
- Not much money is made on Fri
- Lot money is made on Sat

```python
# Horizontal bar plot: When categories names are too long or too many

sns.barplot(
    data=tips,
    y="day",
    x="total_bill"
)

plt.title("Average Total Bill per Day")
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.sample(10)
```

```python
sns.barplot(flights, 
            x="year",
            y="passengers",
            estimator="sum",
            errorbar=None)
plt.show()
```

# Observ:
- Number of passengers have been steadily increasing

```python
# Add text labels with each bar’s value:
ax = sns.barplot(flights,
                 x="year",
                 y="passengers",
                 estimator="sum",
                 errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)
plt.show()
```
