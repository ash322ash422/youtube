# stripplot


A strip plot displays individual data points for a numerical variable across categories.

- Strip plot shows every single observation.
- Do NOT use Strip Plot for very large datasets

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
sns.stripplot(
    data=tips,
    y="total_bill"
)

plt.title("Strip Plot of Total Bill")
plt.show()
```

# Observ:
- Each dot is a data point
- Most of data is between 10 and 25

```python
# compare with boxplot

# Create figure with 2 columns
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Stripplot
sns.stripplot(
    data=tips,
    y="total_bill",
    ax=axes[0]
)
axes[0].set_title("Strip Plot")

# Boxplot
sns.boxplot(
    data=tips,
    y="total_bill",
    ax=axes[1]
)
axes[1].set_title("Box Plot")

plt.tight_layout()
plt.show()
```

# Observ: 
- 25% of data lies below 14(approx)
- 50% of data lies below 18(approx)
- 75% of data lies below 24(approx)


```python
# Now lets add category day

sns.stripplot(
    data=tips,
    x="day", # category
    y="total_bill"
)

plt.title("Strip Plot of Total Bill per Day")
plt.show()
```

Each dot = one customer

Vertical spread = distribution

# Observation:
- Not many custoomers on Fri



```python
# Add jitter to prevent points overlap.

sns.stripplot(
    data=tips,
    x="day", # category
    y="total_bill",
    jitter=True
)

plt.title("Strip Plot with Jitter")
plt.show()
```

```python
# jitter
sns.stripplot(
    data=tips,
    x="day", # category
    y="total_bill",
    hue="time", # category
    jitter=True
)

plt.title("Total Bill per Day by time")
plt.show()
```

# observ:
- All customers on Sat and Sun are for Dinner
- More customers for Lunch on Thur

```python
# or seperate the points
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time", # category
    dodge=True
)

plt.title("Total Bill per Day by time")
plt.show()


```
