# boxplot()


A **boxplot (box-and-whisker plot)** summarizes the **distribution of numerical data** using **five key statistics**:

1. Minimum
2. First Quartile (Q1 – 25%)
3. Median (Q2 – 50%)
4. Third Quartile (Q3 – 75%)
5. Maximum


**Middle 50% of the data lives inside the box**

## Why do we need a Boxplot?

### 1 Shows distribution at a glance

* Where is the data centered?
* How spread out is it?
* Is it skewed?

 Much faster than inspecting raw numbers.

---

### 2️ Identifies outliers clearly

* Points beyond whiskers = **outliers**
* Indicate unusual or extreme values


---

### 3️ Excellent for comparison

* Compare distributions **across categories**
* Much better than histograms when comparing groups


### 🔹 Median Line (Q2)

* Line inside the box
* Shows the **central value**
* If not centered → data is skewed


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set global figure size (width, height in inches)
plt.rcParams["figure.figsize"] = (8, 4)
```

```python
tips = sns.load_dataset("tips")

tips.sample(10)
```

```python
# 1) Lets do box plot of tip 
sns.boxplot(data=tips,
            x='total_bill'  # Should be numeric
) 

plt.show()
```

# Observation
- Q1 = 13 (approx). 25% of total bill data is < 13
- Q2 = 17 (approx). 50% of total bill data is < 17
- Q3 = 24 (approx). 75% of total bill data is < 24
- 50% of customers payed total bill between 13 and 24 approx.
- Few customers (outliers) payed high total bill > 40
- Median (Q2) is not in the center. So the data is skewed.

```python
# 2) vertical boxplot
sns.boxplot(data=tips,
            y='total_bill'  # Should be numeric
) 

plt.show()
```

# Numeric value categorized

```python
# Lets do box plot of total bill with categorized by size

sns.boxplot(data=tips,
            x='size',       # Should be category,
            y='total_bill'  # Should be numeric
) 

plt.show()
```

# Observation:
- As size goes up, the range of total bill goes up.
- 

```python
#  Lets do box plot of tip with categorized by day

sns.boxplot(data=tips,
            x='day',       # Should be category,
            y='tip'  # Should be numeric
) 

plt.show()
```

# Observation:
- 50% of customers on Thur payed tip between 2 and 3.6
- 50% of customers on Sun payed tip amount between 2 and 4.1
- On Thur and Sat, few customers payed higher tips (Outliers)
- More tips on Sun

```python
# 2) now we add sex 

sns.boxplot(data=tips,
            x='day',  # category,
            y='tip',  # numeric
            hue='sex' # category
) 

plt.show()
```

# observation: 
- males are paying higher tips than female

```python
# 3) fill = False

sns.boxplot(data=tips,
            x='day',  # category,
            y='tip',  # numeric
            hue='sex', # category
            fill= False
) 

plt.show()
```

```python
# lets see total_bill categorized by time
sns.boxplot(data=tips,
            x='day', # category
            y='total_bill', # numerical
            hue = 'time') 

plt.show()
```

# observation: 
- on fri people come for both lunch and dinner.
- on fri, total bill is higher for dinner than for lunch
- not many people come for dinner on thur
- on sat and sun, people come for dinner mainly


# skewed

```python
thursday_tips = tips[tips["day"] == "Thur"]
thursday_tips
```

```python
# histogram and KDE
sns.histplot( data=thursday_tips,
              x='tip',
              bins=10,
              kde=True
)
plt.show()

# Skewed towards right
```

```python
# Compare with boxplot
sns.boxplot(data=thursday_tips,
            x='tip'  # Should be numeric
) 

plt.show()
```

## Observation about skewed data:
- The median is shifted towards left
- The right whisker line is longer than left whisker line
